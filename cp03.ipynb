## Introdução

Neste capítulo, vamos analisar as decisões de consumir e poupar. 

Para analisar esse problema de forma eficiente, precisamos de um mínimo de teoria para organizar as idéias. Teorias e modelos são tentativas de obter respostas para perguntas sobre como o mundo funciona, são tentativas de resolver problemas. Se não temos perguntas, não precisamos de teoria. Nossa pergunta inicial é bastante simples: como as pessoas decidem quanto dos seus recursos vão poupar ou consumir? 

Para responder à pergunta acima – e a outras –, nossa abordagem terá dois elementos principais que serão detalhados mais adiante: (i) analisar o problema no ambiente mais simples possível; (ii) basear a análise nas hipóteses de otimização de objetivos individuais e de equilíbrio de mercado.

Podemos, de forma esquemática, pensar no seguinte roteiro de análise:

(i) Definição do problema a ser analisado. 

(ii) Definição das hipóteses simplificadoras e representação matemática do problema. 

(iii) Solução do modelo: cálculo dos comportamentos ótimos individuais, do equilíbrio da economia e da reação da economia aos choques.

(iv) Interpretação dos resultados e avaliação empírica.

Esse roteiro, altamente esquemático, indica as principais etapas da análise. Começamos com a definição do problema, construímos um modelo simplificado baseado em hipóteses iniciais, resolvemos o modelo e avaliamos os resultados. Caso necessário, é possível estender o modelo para discutir questões adicionais. Isso implica refazer o roteiro de análise, aproveitando os resultados apropriados.


## As hipóteses simplificadoras

Iniciantes costumam pensar que para fazer uma análise apropriada de um problema é necessário incluir na teoria a maior quantidade possível de dados da realidade. Quanto mais detalhista, mais realista e, portanto, melhor. Um pouco de reflexão indica, entretanto, que essa não é uma boa estratégia de análise. 

Pensemos no problema de consumo e poupança. Será que para entendermos o problema é importante saber os nomes das pessoas que tomam essas decisões? Qual seu time de futebol preferido? Essas perguntas podem parecer tolas, mas elas indicam que tentar construir uma teoria com base em uma descrição detalhista do mundo não é uma boa estratégia. Na verdade, parece uma estratégia inviável: onde ela termina? 

Uma estratégia mais promissora vai no sentido oposto. Vamos começar tentando eliminar da discussão tudo aquilo que não pareça indispensável para entender o problema. 

Para fazer uma análise organizada, precisamos abstrair das características secundárias do problema para concentrar apenas no essencial. Teoria e modelos são simplificações, caricaturas, da realidade. As simplificações podem ser úteis, concentrando a atenção no essencial, ou enganosas, desviando a atenção do essencial e induzindo ao erro. A mesma simplificação pode ser útil ou enganosa dependendo do problema a ser analisado. 

Por exemplo, consulte alguns mapas do Brasil. Só existe um Brasil, mas há muitos mapas diferentes: divisão por estados, por clima, densidade populacional... Qual está certo? É evidente que a pergunta não faz sentido. Os mapas retratam aspectos diferentes da realidade e, por isso, fazem simplificações diferentes. Todos podem ser úteis para seus propósitos. Um mapa que tentasse capturar todos os aspectos da realidade seria complicado demais para ser útil. Mas isso não quer dizer que “vale tudo”. Por exemplo, um mapa do Brasil que coloque o Paraná ao lado da Bahia.

Modelos, assim como mapas, são simplificações. Dependendo do objetivo, a mesma simplificação pode ser útil ou enganosa.

### As preferências

Inicialmente, vamos analisar a decisão de consumo e poupança de um indivíduo que recebe uma quantidade de recursos exógena em cada um dos dois únicos períodos de sua vida. O bem-estar do indivíduo só depende de quanto ele consome ao longo da vida e o problema econômico que ele deve resolver é como alocar os recursos de que dispõe entre consumo “hoje” $(c_1)$ e consumo “amanhã” $(c_2)$. 

Intuitivamente, parece razoável que a pessoa não queira consumir todos os seus recursos em único período, “passando fome” no outro período. A análise formal desse problema é simples, mas servirá de base para problemas mais avançados.
Vamos supor que o bem-estar do indivíduo ao longo da vida possa ser representado pela seguinte função de utilidade:

$U(c_1,c_2) = u(c_1) + \cfrac{u(c_2)}{1+ρ} = ln(c_1 ) + \cfrac{ln(c_2)}{1+ρ} $

Em palavras, a utilidade total do indivíduo ao longo da vida $(U(c_1,c_2))$ é uma soma ponderada das utilidades instantâneas $(u(c_1),u(c_2))$ que o indivíduo obtém por consumir a cada período. A letra grega rho (ρ≥0) representa a impaciência do indivíduo. Se $ρ>0$, o indivíduo prefere consumir uma dada quantidade hoje ao invés da mesma quantidade amanhã. 

A segunda igualdade utiliza como função de utilidade específica a função logarítmica, uma especificação que usaremos com frequência. Outras funções de utilidade serão vistas mais adiante. Em todos os casos suporemos que:

$ \cfrac{du(c)}{dc} >0  \,\,\,\,\,\,\,\,\,\,\,\,\,\,   \cfrac{d^2 u(c)}{dc^2} < 0 $
   
Onde as desigualdades utilizam o fato de que as quantidades consumidas não podem ser negativas. A primeira hipótese (utilidade marginal positiva) significa que o indivíduo é insaciável e sempre prefere mais a menos. Essa hipótese parece razoável para a larga maioria das pessoas, pois poucas diriam que já tem o suficiente de tudo. Mas ela poderia não ser razoável se estivéssemos considerando bens de consumo específicos ou indivíduos com preferências bastante diferentes das habituais. A segunda hipótese (utilidade marginal decrescente) indica que quantidades adicionais de consumo trazem ganhos cada vez menores de bem-estar.

No caso da função logarítmica:

$ \cfrac{du(c)}{dc} = \cfrac{dln(c)}{dc} = \cfrac{1}{c} > 0 \,\,\,\,\,\,\,\,\,\,\,\,\,\,
   \cfrac{d^2 u(c)}{dc^2} = \cfrac{d^2 ln(c)}{dc^2} = -\cfrac{1}{c^2} < 0 $



Vale a pena chamar a atenção para algumas hipóteses implícitas na especificação adotada para a função de utilidade. Estamos supondo que as preferências do indivíduo possam ser representadas por uma função de utilidade que: (i) depende apenas do consumo; (ii) não muda ao longo da vida; e (iii) é separável intertemporalmente. O que isso significa?

(i) É claro que muitas coisas além do consumo afetam o bem-estar das pessoas, mas elas não são indispensáveis para nossa análise da decisão de consumo e poupança. Um elemento importante que está ausente da função de utilidade é a quantidade de lazer de que o indivíduo desfruta. A razão para essa omissão é que supusemos que a renda do indivíduo seja exógena. Mais adiante, quando discutirmos a decisão de quantas horas trabalhar, incluiremos o lazer na função de utilidade. 

(ii) Utilizarmos a mesma função de utilidade instantânea em cada período é claramente uma simplificação. Mas pode ser uma boa aproximação se as preferências não mudarem muito. Note que estamos nos referindo ao consumo em geral e não a bens de consumo específicos. Por exemplo, crianças e adultos preferem bens diferentes, mas ambos preferem ter mais consumo a ter menos.

(iii) Supomos que o consumo em um período só gere satisfação naquele período. A satisfação ao consumir hoje não depende de quanto foi consumido ontem ou será consumido amanhã. Dessa forma, podemos representar a felicidade total do indivíduo ao longo da vida como a soma (ponderada) de sua felicidade em cada período. Ou seja, a utilidade total é uma [função separável](https://calculus.subwiki.org/wiki/Additively_separable_function) das utilidades em cada período. 

A separabilidade intertemporal está longe de ser uma característica evidente das preferências das pessoas. Ela implica, por exemplo, que a minha satisfação em jantar não depende de eu ter almoçado ou não. É claro que, nesse caso, a hipótese é pouco plausível. Ela se torna mais razoável se estivermos falando do consumo em períodos distantes entre si. Na verdade, a hipótese de separabilidade intertemporal é feita principalmente para simplificar a análise, mas pode não ser adequada para alguns problemas. Por exemplo, ela não seria adequada para a análise de hábitos de consumo.


### O objetivo individual

Vamos supor que o objetivo do indivíduo seja obter o máximo de satisfação possível ao longo da vida. Ou seja, o indivíduo fará suas escolhas de consumo e poupança para maximizar sua utilidade intertemporal $ \left( U(c_1,c_2) \right) $.

Assumimos, portanto, que os indivíduos tentem, e consigam, fazer aquilo que é melhor para si próprios, dentro de suas possibilidades. A primeira parte parece óbvia: as pessoas tentam conseguir aquilo que lhes parece ser a melhor alternativa. Mas será que é razoável supor que elas consigam?

Maximização é uma alternativa simples para capturar a tentativa das pessoas de fazer o melhor para elas próprias. Interpretada realisticamente, não significa que as pessoas sempre conseguirão fazer as escolhas ideais, alocando perfeitamente seu orçamento até o último centavo. Significa que os indivíduos não irão ignorar sistematicamente maneiras claras de melhorar suas próprias vidas. 

Podemos comparar o comportamento otimizador com a situação de um motorista que tenta dirigir mantendo o carro entre as faixas de trânsito. Ele não conseguirá fazer isso perfeitamente o tempo todo. Mas se o carro começar a se afastar demais para um dos lados da pista, o motorista irá corrigir a trajetória. Seria difícil prever todas essas pequenas alterações de curso. A trajetória ótima, idealmente sempre no centro das faixas, é uma maneira simples de descrever aproximadamente a trajetória real.


### As restrições 

Os agentes têm uma restrição orçamentária para cada um dos dois períodos. 

(i) Em cada período, eles recebem uma renda exógena $(y_1,y_2)$. 

(ii) Os agentes podem emprestar ou pedir emprestado, comprando ou vendendo títulos $(a)$ com duração de um período; $a>0$ significa que o agente é credor e $a<0$ indica que é devedor. Os títulos comprados ou vendidos "ontem" $(a_0)$ pagam uma taxa $r_0$ e são resgatados "hoje" pelo valor, positivo ou negativo, $(1+r+0)a_0$. Essas transações foram decididas no passado e não podem ser alteradas. "Hoje" o agente compra ou vende títulos no valor $a_1$, à taxa $r_1$, que serão resgatados "amnhã" por $(1+r+1)a_1$.

(iii) Os gastos de consumo são $(c_1,c_2)$.

Dessa forma, as restrições orçamentárias para os dois períodos são:

$  \left\{ \begin{array}{l l}	
     y_1 + (1+r_0)a_0 = c_1 + a_1 \\
     y_2 + (1+r_1)a_1 = c_2 + a_2  \\
    \end{array}\right. $
 
A restrição para o primeiro período indica que o total de recursos disponíveis para o agente no primeiro período (lado esquerdo da equação) deve ser igual ao dispêndio total nesse período (lado direito da equação). A restrição orçamentária para o segundo período tem uma interpretação equivalente.

A restrição do primeiro período pode ser reescrita como:

$ y_1 + r_0a_0 = c_1 + (a_1 - a_0) = c_1 + s_1 $
     
Ou seja, o fluxo de renda do primeiro período (renda exógena mais remuneração do patrimônio; lado esquerdo) deve ser igual ao consumo mais a poupança no mesmo período. E, de forma equivalente, para o segundo período.

Os patrimônios $(a_0,a_1,a_2)$ são estoques. As poupanças $(s_1=a_1-a_0)$  e $(s_2=a_2-a_1)$, que correspondem às variações dos patrimônios, são fluxos.

As restrições orçamentárias para os dois períodos podem ser consolidadas em uma única restrição orçamentária intertemporal. Substitua $a_1$ da restrição de um período na restrição do outro período e reorganize. O resultado é (onde *w* é a riqueza, o valor presente do fluxo de recursos):

$ w ≡ y_1 + (1+r_0).a_0 + \cfrac{y_2}{1+r_1} =c_1 + \cfrac{c_2}{1+r_1} + \cfrac{a_2}{1+r_1} $
 
A última equação é conhecida como a restrição orçamentária intertemporal do indivíduo e indica que o valor presente do seu fluxo de renda deve ser igual ao valor presente do seu fluxo de gastos.  

**Exemplo 1** Suponha: $ y_1=100, \,\, y_2=110, \,\, r = 0.1 $. Então:

$ w = 100 + \cfrac{110}{1.1} = 200 $     

♣

Finalmente, suporemos que o patrimônio final do indivíduo não possa ser negativo $(a_2≥0)$. Isso significa que os indivíduos podem deixar herança, mas não podem deixar dívidas.  

Por que os agentes não podem deixar dívidas? Porque estamos considerando um ambiente extremamente simplificado em que não há incerteza. Sendo assim, tanto quem pede recursos emprestados quanto quem empresta sabe se o tomador do empréstimo irá ou não pagar. Nesse caso, é claro que as pessoas que não irão reembolsar os empréstimos não os obterão. Se quisermos introduzir a possibilidade de não pagamento das dívidas de maneira razoável, precisaremos de um ambiente com incerteza. Dito de outra forma, nosso modelo simplificado precisa ser utilizado com muita cautela para discutir calotes de dívidas.



### As variáveis de escolha

O indivíduo tem quatro variáveis de escolha: seus níveis de consumo $(c_1,c_2)$ e de patrimônio $(a_1,a_2)$ . O objetivo do consumidor é escolher 
essas quatro variáveis para maximizar o valor de sua utilidade total ao longo da vida, sujeito às restrições orçamentárias. Portanto, as quatro escolhas não são independentes.

Além disso, os agentes são tomadores de preços (*price takers*). Ninguém é (economicamente) grande o bastante para, 
individualmente, influenciar os preços de mercado. No nosso modelo, há dois bens. O bem hoje e o bem amanhã. A taxa de juros 
é o preço relativo desses bens. Uma unidade do bem hoje vale   amanhã. Em outras palavras, o bem hoje é o numerário 
(unidade de conta) dessa economia. Esse é o único preço relativo desse modelo. Estamos supondo, portanto, que a taxa 
de juros é determinada pelo mercado e não é afetada pelas decisões de um indivíduo específico. 


### O modelo matemático

Um agente econômico é definido pelo problema econômico que ele resolve. Ou seja, o agente é definido por: (i) sua função objetivo; (ii) suas variáveis de controle (variáveis de escolha); (iii) sua dotação inicial de recursos; (iv) suas restrições: orçamentárias, tecnológicas, de informação... 

Dadas as hipóteses feitas acima, o problema do indivíduo pode ser representado matematicamente da seguinte forma.

Dadas as hipóteses feitas acima, o problema do indivíduo pode ser representado matematicamente da seguinte forma.

$$ A \left\{ \begin{array}{c l}	
\underset{(c_1,c_2,a_1,a_2)}{max} U(c_1,c_2)  = u(c_1) + \cfrac{u(c_2)}{1+\rho}\\
     y_1 + (1+r_0)a_0 = c_1 + a_1 \\
     y_2 + (1+r_1)a_1 = c_2 + a_2  \\
     A_0:exógeno   \\
     c_1,c_2,a_2 \ge 0  \\
     u'>0 \;, \; u"<0   \\
     \lim \limits_{c\to 0}  u'(c) = \infty \; ,   \;\;\; \lim \limits_{c\to \infty}  u'(c) = 0
\end{array}\right. $$

Para facilitar a análise vamos, inicialmente, trabalhar com uma versão simplificada do problema A. Vamos supor que a função de utilidade instantânea seja logarítmica $ (u(c)=ln(c)) $ e que os indivíduos comecem e terminem a vida sem ativos financeiros $(A_0=A_2=0)$. 

As versões simplificadas do problema são:

$$ B \left\{ \begin{array}{c l}	
\underset{(c_1,c_2,a_1)}{max} U(c_1,c_2)  = u(c_1) + \cfrac{u(c_2)}{1+\rho}\\
     y_1 = c_1 + a_1 \\
     y_2 + (1+r)a_1 = c_2  \\
    \end{array}\right. $$

$$ C \left\{ \begin{array}{c l}	
\underset{(c_1,c_2)}{max} U(c_1,c_2)  = u(c_1) + \cfrac{u(c_2)}{1+\rho}\\
     y_1 + \cfrac{y_2}{1+r} = c_1 + \cfrac{c_2}{1+r}  \\
    \end{array}\right. $$

### Hipóteses implícitas

Vale a pena chamar a atenção para algumas hipóteses implícitas na formulação acima.

(i) Um único bem (perecível).

A decisão de poupança envolve a troca entre o consumo de bens ou serviços hoje – não importam quais – e o consumo de bens ou serviços amanhã – não importam quais. Uma pessoa pode deixar de ir ao cinema hoje para jantar fora no próximo final de semana. Ou pode decidir comprar um computador hoje sabendo que terá que desistir de viajar no final do ano. O que ela obtém e do que ela abre mão não importam para nossa análise inicial. Queremos entender o elemento comum às duas situações: em ambos os casos, a pessoa abre mão de consumo em algum momento em troca de consumo em outro momento. Essa simplificação é útil para a discussão atual, mas poderia não ser apropriada para outras. Por exemplo, ela não seria apropriada se quiséssemos discutir mudanças de preços relativos; por exemplo, preço da alimentação frente aos demais bens de consumo.

Suporemos que esse único bem “genérico” é perecível – não pode ser armazenado. A discussão sobre o consumo de bens duráveis é um pouco mais difícil e pode ser deixada para mais tarde.

(ii) Ausência de incerteza

A incerteza quanto ao futuro certamente é um elemento importante nas decisões de poupança. Mas as pessoas poupam também por razões que (praticamente) independem da incerteza. Por exemplo, a decisão de poupar para viajar nas férias de final de ano não é motivada principalmente por elementos de incerteza. Problemas envolvendo risco são mais complicados que os sem risco. Para simplificar, vamos começar a análise pelos problemas sem incerteza.

(iii) Renda exógena.

Queremos analisar a utilização dos recursos disponíveis, independente de como são obtidos. Nesse caso, podemos supor, simplificadamente, que as pessoas recebam os bens de consumo gratuitamente, sem precisar trabalhar. É claro que isso não representa a situação da larga maioria das pessoas (adultas) no mundo real. Essa simplificação é útil para separarmos, conceitualmente, a decisão de consumo e poupança da decisão de trabalhar e da decisão de acumular ativos produtivos. 

(iv) Sistema financeiro simplificado.

Há um único ativo financeiro que paga uma taxa de juros fixa e pré-fixada.
Não há spread bancário: as pessoas podem emprestar e pedir emprestado à mesma taxa de juros. (É claro que os bancos não funcionam assim.)

Não há restrição de liquidez. O indivíduo pode pedir emprestado tanto quanto queira desde que tenha recursos para pagar seus empréstimos. Isso significa que não há um limite inferior para o valor dos ativos financeiros.

(v) Estrutura do mercado

Economia competitiva, ou seja, os indivíduos são pequenos demais para afetar os preços de mercado. A economia pode ser fechada (autarquia) ou uma pequena economia aberta, incapaz de afetar os preços internacionais. 

Não há governo. Não há ativos monetários: não há preços nominais, apenas relativos.

(vi) Um único tipo de indivíduo.

Dependendo do ponto de vista, as pessoas podem ser consideradas como sendo muito parecidas ou muito diferentes. A maioria das pessoas prefere consumir muito que consumir pouco, mas elas preferem bens de consumo diferentes. O que nos interessa é analisar a decisão de consumir mais ou menos, não importando o que vai ser consumido. Desse ponto de vista, as pessoas são bastante parecidas, no sentido de desejarem a maior quantidade de consumo possível. Portanto, vamos supor que exista um agente representativo na economia. 

Essa simplificação poderia não ser útil se, por exemplo, quiséssemos entender porque algumas pessoas investem em ativos de alto risco e outras em ativos de baixo risco. Ou se quiséssemos discutir problemas de distribuição de renda. Nesses casos, seria necessário levar em conta as diferenças entre as pessoas.


## Conclusão

    A scientific theory should be as simple as possible, but no simpler. (Albert Einstein, diversas versões.)

Fizemos uma longa lista de simplificações. Várias serão eliminadas nos próximos capítulos quando não forem mais úteis. Como avaliar essas hipóteses simplificadoras? As simplificações não são, intrinsecamente, certas ou erradas. Elas podem ser úteis ou enganosas para o problema que queremos analisar. Hipóteses úteis para a análise de um problema podem ser enganosas para a análise de outro problema. Para chegarmos a uma conclusão, precisaremos comparar o funcionamento da nossa economia simplificada com o funcionamento das economias reais. Como a lista de simplificações é longa, uma pergunta natural é: será que um modelo tão simplificado ainda pode ser útil para alguma coisa? 

Lembre que um modelo é uma representação deliberadamente simplificada da realidade. As simplificações têm como objetivo dar destaque àquilo que é fundamental para analisar o problema que está sendo discutido e deixar de lado aquilo que é secundário para aquele problema. Problemas diferentes requerem simplificações – modelos – diferentes.

Por exemplo, no nosso caso, estamos interessados no problema da utilização dos recursos disponíveis para consumo corrente ou futuro. É claro que uma das razões que leva as pessoas a pouparem é a incerteza quanto ao futuro. Essa, entretanto, não é a única razão. Mesmo que sua incerteza quanto ao futuro seja muito pequena, talvez você queira poupar hoje para comprar um bem ou serviço caro no futuro: um apartamento, uma viagem de férias... Ao supormos, simplificadamente, que não haja incerteza no modelo, não estamos dizendo que não haja incerteza no mundo ou que ela não seja importante. Estamos, apenas, tentando entender comportamentos cuja motivação principal não é a incerteza. O mesmo se aplica às demais hipóteses simplificadoras.

Novamente, será que um modelo tão simples quanto o apresentado aqui pode ser útil? A resposta dependerá da capacidade do modelo em explicar aquilo que estamos tentando entender. É preciso confrontar as implicações do modelo com aquilo que observamos. Se as conclusões do modelo conflitarem com as observações e se não houver erros de lógica, alguma hipótese que parecia apenas uma simplificação estará afetando negativamente os resultados. A capacidade de explicar a realidade é o teste definitivo para qualquer teoria.

Em certo sentido, não há problemas exclusivamente econômicos. Há componentes econômicos dos problemas da vida das pessoas. Isso não é apenas uma questão retórica. Existem inúmeras razões – culturais, psicológicas, históricas... – que afetam as decisões das pessoas. Existem também, motivações econômicas. Essas são as que os economistas estudam. Em alguns problemas, a motivação econômica pode ser preponderante (alocação de portfólio), em outros, secundária (com quem casar). O fato de termos um modelo tão simplificado deveria nos prevenir contra uma aplicação mecânica de suas conclusões em situações em que suas hipóteses em vez de simplificadoras sejam enganosas.

Há algumas vantagens em começarmos com um modelo muito simples – desde que não seja enganoso. 

Primeiro, é mais fácil de entender. Segundo, é mais fácil perceber quais características do mundo real afetam quais tipos de comportamento. Por exemplo, se conseguirmos estabelecer uma razão para as pessoas pouparem em um modelo sem incerteza, teremos identificado um componente do comportamento das pessoas que existirá em ambientes de alta ou baixa incerteza. Isso ajuda a organizar as ideias. Terceiro, o modelo simples pode servir de base para a construção de modelos mais complexos. 

Quando devemos passar de um modelo mais simples para outro mais complexo? A resposta curta é: quando for indispensável; quando for impossível analisar um problema com a estrutura mais simples. Por exemplo: um modelo com um único bem pode ser útil para discutir problemas de consumo e poupança, mas é inútil para analisar os efeitos de uma desvalorização cambial – com apenas um bem, não há preço relativo e não há taxa de câmbio. 

Embora problemas diferentes possam requerer hipóteses simplificadoras diferentes, a abordagem que utilizaremos será sempre a mesma e seguirá o roteiro apresentado acima. Em particular, adotaremos sempre a abordagem que consiste em representar o problema econômico no ambiente mais simples possível, utilizando as hipóteses de otimização de objetivos individuais e de equilíbrio de mercado.

### A função de utilidade 

A função de utilidade logarítmica é um caso particular da função de elasticidade de substituição constante (CES): 


$ u(C)=\left\{ \begin{array}{c l}
 \cfrac{c^{1-\theta}-1}{1-\theta} \;, \;\;\;\; \theta \neq 1 \\
 ln(c) \;, \;\;\;\; \theta=1  \\
    \end{array}\right. $

A expressão acima indica que, à medida que teta se aproxima de 1, a função CES fica cada vez parecida com a função log. Podemos verificar esse resultado analiticamente notando que para $ \theta = 1$, a função CES não está definida e aplicando [L'Hôpital](https://pt.khanacademy.org/math/calculus-home/derivative-applications-calc/lhopitals-rule-calc/a/lhopitals-rule-review) :

$ \cfrac{d}{d \theta} \left(\cfrac{c^{1-\theta}-1}{1-\theta} \right) = 
\cfrac{c^{1-\theta}(-1)ln(c)}{-1} = c^{1-1}ln(c) = ln(c) $

Também podemos ilustrar o resultado graficamente fazendo o gráfico da CES para valores de teta cada vez mais próximos de 1.

A próxima célula calcula o valor da riqueza dadas as rendas e as taxas de juros em um número arbitrário de períodos. O código segue os seguintes passos:

(i) Define uma função WN que tem como argumentos os vetores y e r. y e r devem ter o mesmo tamanho, mas esse tamanho é arbitrário. Essa função começa em function e termina no último end.

(ii) Verifica o tamanho do vetor y.

(iii) Calcula um valor provisório para os fatores de desconto R.

(iv) Define um loop para calcular os valores definitivos de R. O loop começa em for e termina no primeiro end.

(v) Estabelece valores para y e r.

(vi) Chama a função WN com os valores estabelecidos para os parãmetros.


````julia
function WN(y,r)                 # início da função WN

    N = length(y)
    R = 1.0./(1.0.+r)            # valores provisórios para R
 
    for k = 2:N
        R[k] = R[k].*R[k-1]      # valores definitivos para R
    end
        
    WN = y*R'
    
    end                         # final da função WN

y = [100 220 121]
r = [0 0.1 0.1]

WN(y,r)
````
