## Introdução

Os ultimos dois capítulos apresentaram o problema do consumidor e as condições de ótimo. Neste capítulo, veremos o equilíbrio da economia e alguns exemplos numéricos.

Vamos analisar o equilíbrio da economia em duas situações extremas. Pequena economia aberta e autarquia (economia fechada). 
Na pequena economia aberta, consideramos que o país é pequeno, do ponto de vista econômico, e, portanto, incapaz de afetar os preços no mercado internacional. No nosso modelo, há apenas um preço relativo, a taxa de juros. Portanto, a pequena economia aberta pode emprestar e pedir emprestado no exterior tanto quanto queira a uma taxa de juros fixa e determinada no exterior.
Na autarquia, obviamente, não há transações com o exterior. Portanto, o valor agregado dos ativos financeiros deve ser zero: os créditos e débitos mútuos no país se anulam. A taxa de juros, por outro lado, será endógena, determinada no mercado doméstico.

## Equilíbrio na pequena economia aberta

Na pequena economia aberta, a taxa de juros doméstica é igual à taxa internacional. Como a economia é pequena no mercado internacional, os indivíduos podem, agregadamente, transacionar quanto queiram com exterior sem alterar a taxa de juros internacional. Ou seja, a taxa de juros doméstica é exógena em equilíbrio. 

Note que, quando analisamos o problema do indivíduo supusemos que a taxa de juros fosse exógena para ele, individualmente, pois ele é pequeno demais para afetar os juros de mercado. Não havíamos dito nada sobre a taxa de juros de equilíbrio. Agora, estamos supondo que a economia como um todo é pequena no mercado internacional e não afeta a taxa de juros internacional.

Qual a consequência disso para o equilíbrio da economia? Lembre que estamos supondo, simplificadamente, que todos os indivíduos sejam iguais. Suponha que o número de pessoas na economia (N) seja mil.

Suponha que, no primeiro período, tenhamos para o agente representativo: 

$ y_1=100, \,\,\, c_1^*=110, \,\,\, a_1^*=-10 $

Os valores agregados serão:

$ Y_1=1000.y_1=100.000 \,\,\, C_1^*=1000.c_1^*=110.000 \,\,\, A_1^*=1000.a_1^*=10000 $

Portanto, na pequena economia aberta, os valores de equilíbrio são simplesmente um múltiplo dos valores ótimos obtidos na solução do problema do indivíduo.  Por exemplo, o impacto de uma variação na taxa de juros internacional sobre o consumo de equilíbrio do primeiro período é N vezes o impacto sobre o consumo ótimo individual. A análise do equilíbrio na pequena economia aberta é uma “repetição multiplicada” da análise das escolhas ótimas individuais.


### O consumo corrente

A reação do consumo corrente a mudnças nas variáveis exógenas é:

$ \begin{array}{l l}
 \cfrac{\partial C_1}{\partial Y_1} = \cfrac{1+\rho}{2+\rho}  \in (0,1) &
 \cfrac{\partial C_1}{\partial Y_2} = \cfrac{1+\rho}{2+\rho} 
    \cfrac{1}{1+r} \in (0,1) \\ 
    \\
\cfrac{\partial C_1}{\partial r} = - \cfrac{1+\rho}{2+\rho} 
    \cfrac{Y_2}{(1+r)^2} < 0 &    
\cfrac{\partial C_1}{\partial \rho} = \cfrac{1}{(2+\rho)^2} 
    \left( Y_1+\cfrac{Y_2}{1+r} \right) > 0  
 \end{array} 
 \, \\
 $
 
 **Comentário**
 
(i) Para obter esses resultados, basta calcular as derivadas parciais das equações de equilíbrio. 
Veja o <a id='voltar1'></a> [Exemplo 1](#Link5).
 
(ii) Os resultados são bastante intuitivos. Um aumento na renda em qualquer dos períodos, eleva o consumo do primeiro período. Entretanto, o aumento no consumo é sempre menor que o aumento na renda. Ou seja, uma parte do aumento na renda é sempre poupada. Reduções na renda provocam reduções semelhantes no consumo.

Um aumento na impaciência eleva o consumo corrente.

(iii) Um aumento na taxa de juros reduz o consumo corrente. Ao contrário do que possa parecer, esse não é um resultado geral. Ao contrário dos anteriores, ele depende de estarmos utilizando a função de utilidade logarítmica. Veremos, mais adiante, que uma variação na taxa de juros tem um impacto ambíguo sobre o consumo. A razão pela qual o consumo corrente diminui é que a elevação dos juros reduz o valor presente da renda futura; ou seja, o indivíduo fica mais pobre. (Note que se a renda do segundo período for nula $(Y_2=0)$  , alterações na taxa de juros não afetarão o consumo do primeiro período.)

### O consumo futuro

Os resultados para o consumo futuro podem ser discutidos tanto em termos do valor corrente do consumo futuro $(C_2)$ quanto do seu valor presente $\left( \cfrac{C_2}{1+r} \right) $. Vejamos os resultados para o valor presente; os do valor corrente estão no <a id='voltar2'></a> [Exemplo 2](#Link2).

$ \begin{array}{l l}
\cfrac{ \partial \left( \cfrac{C_2}{1+r}\right)} {\partial Y_1} = \cfrac{1}{2+\rho} \in (0,1)  &
    \cfrac{ \partial \left( \cfrac{C_2}{1+r}\right)}{\partial \left( \cfrac{Y_2}{1+r} \right) } 
    =  \cfrac{1}{2+\rho} \in (0,1) \\
\cfrac{ \partial \left( \cfrac{C_2}{1+r}\right)}{\partial r} = 
    - \cfrac{1}{2+\rho} \cfrac{Y_2}{(1+r)^2} < 0  &
\cfrac{ \partial \left( \cfrac{C_2}{1+r}\right)}{\partial \rho} = 
    - \cfrac{1}{(2+\rho)^2} \left( Y_1+\cfrac{Y_2}{1+r} \right) 
\end{array}
\, \\
$ 

Os resultados são semelhantes aos do consumo corrente e dispensam maiores comentários.

### A poupança

Lembre que na economia aberta, a poupança é igual ao saldo em conta corrente $(S=A_1-A_0=A_1)$. A reação da poupança aos choques é:

$ \begin{array}{l l}
\cfrac{\partial A_1}{\partial Y_1} = \cfrac{1}{2+\rho}  \in (0,1) &
\cfrac{\partial A_1}{\partial Y_2} = - \cfrac{1+\rho}{2+\rho} 
    \cfrac{1}{1+r} \in (-1,0)  \\
\cfrac{\partial A_1}{\partial r} = \cfrac{1+\rho}{2+\rho} 
    \cfrac{Y_2}{(1+r)^2} > 0 &
\cfrac{\partial A_1}{\partial \rho} = \cfrac{1}{(2+\rho)^2} 
    \left( -Y_1+\cfrac{Y_2}{1+r} \right)  
\end{array}
\, \\
$ 

Um aumento na renda corrente aumenta a poupança, pois o indivíduo irá transferir parte desse aumento de renda para o futuro. Uma redução na renda futura tem o mesmo efeito. Ou seja, se você antecipa tempos difíceis no futuro, você começa a se preparar hoje, não espera passivamente a chegada dos tempos de dificuldade. Um aumento na impaciência faz com que o indivíduo queira consumir mais no presente e, portanto, reduz a poupança. 

Finalmente, um aumento na taxa de juros aumenta a poupança. É importante entender a razão para esse resultado.  A expressão para a poupança mostra que os juros só afetam a poupança através do valor presente da renda futura. Quando os juros aumentam, o valor presente da renda futura diminui e o indivíduo poupa mais para transferir recursos para o futuro. Se a renda futura for zero, a variação nos juros não afeta a poupança. Esse não é um resultado geral. Ele depende de estarmos utilizando a função de utilidade logarítmica, em que as propensões a consumir e a poupar não dependem da taxa de juros. 

## Equilíbrio na autarquia

Em uma economia fechada, obviamente não há transações com o exterior. Dessa forma, o total dos ativos financeiros agregados é zero – créditos e débitos mútuos se cancelam:  $ A_1=\sum_{i=1}^{N}a_1^i=0$. Como todos os indivíduos são iguais, isso implica:$a_i=0$. Ou seja, em equilíbrio, ninguém deseja emprestar ou pedir emprestado.

Efetivamente, como todos os agentes são iguais, se um quiser pedir emprestado, todos os demais também quererão e não haverá ninguém para emprestar. Da mesma forma, se um quiser emprestar, todos quererão e não haverá ninguém para pedir emprestado. A única possibilidade é que ninguém empreste ou pegue emprestado. Portanto, agregadamente na autarquia: $C_1=Y_1 ,C_2=Y_2$ . 

Ao tomar sua decisão individual, o agente supõe que, à taxa de juros de mercado, ele pode emprestar ou pedir emprestado tanto quanto queira – desde que seja solvente. Isso implica que a taxa de juros de equilíbrio deverá ser tal que os agentes escolham não emprestar ou pedir emprestado. Como isso ocorre? 

Suponha que, à taxa de juros inicial o indivíduo deseje pedir emprestado. Ele pode fazer isso. Mas como todos estarão fazendo o mesmo, haverá excesso de demanda por empréstimos e a taxa de juros subirá até que o excesso de demanda seja eliminado. Da mesma forma se, à taxa de juros inicial o indivíduo desejar poupar, haverá excesso de oferta de recursos e a taxa de juros cairá. O único equilíbrio possível é quando a taxa de juros for tal que ninguém deseje emprestar ou pedir emprestado. (É claro que se houvesse diferenças entre os indivíduos, em equilíbrio alguns indivíduos poderiam emprestar e outros pedir emprestado.)

Qual será a taxa de juros de equilíbrio? Lembre que em equilíbrio: (i) oferta igual a demanda em cada mercado; (ii) as condições de ótimo individuais devem ser satisfeitas. Ou seja:

$ \left\{ \begin{array}{l l} 
\cfrac{C_2}{C_1} = \cfrac{1+r}{1+ρ} \\
C_1=Y_1 , \\
C_2=Y_2 
\end{array}\right. 
\, \\
$

Portanto, em equilíbrio:

$ 
\cfrac{Y_2}{Y_1} =  \cfrac{1+r}{1+ρ} \Rightarrow r= (1+ρ)\cfrac{Y_2}{Y_1}-1 
\, \\
$

Se, por exemplo, as rendas forem iguais, a taxa de juros de equilíbrio na autarquia será igual à taxa de impaciência.

Um aumento na renda futura, tudo o mais constante, eleva a taxa de juros. A razão é que quando a renda futura aumenta, os indivíduos desejam transferir parte desses recursos para o presente. Para isso aumentam sua demanda por empréstimos, elevando a taxa de juros. Quando a renda corrente aumenta, tudo o mais constante, os indivíduos desejam poupar parte dos recursos, aumentando a oferta de crédito e derrubando a taxa de juros. Variações no gasto público têm efeito oposto ao de variações na renda.

Portanto, a reação da economia aos choques é:

$ \begin{array}{l l}
 \cfrac{\partial C_1}{\partial Y_1} = 1 &
 \cfrac{\partial C_1}{\partial Y_2} = 
     \cfrac{\partial C_1}{\partial r} = 
\cfrac{\partial C_1}{\partial \rho} = 0  
 \end{array} 
 \, \\
 $
 
 $ \begin{array}{l l}
 \cfrac{\partial C_2}{\partial Y_2} = 1 &
 \cfrac{\partial C_2}{\partial Y_1} = 
     \cfrac{\partial C_2}{\partial r} = 
\cfrac{\partial C_2}{\partial \rho} = 0  
 \end{array} 
 \, \\
 $
 
 $ \begin{array}{l l}
 \cfrac{\partial r}{\partial Y_1} = - (1+\rho)\cfrac{Y_2}{Y_1^2} < 0 &
 \cfrac{\partial r}{\partial Y_2} = \cfrac{1+\rho}{Y_1} > 0 &
     \cfrac{\partial r}{\partial \rho} = \cfrac{Y_2}{Y_1} > 0 
 \end{array} 
 \, \\
 $



## Equilíbrio com governo

No último capítulo vimos como uma política fiscal simplificada alterava o problema do consumidor. Agora vamos ver os impactos sobre o equilíbrio. Suporemos uma situação simples em que o orçamento esteja equilibrado a cada período: 

$G_1=T_1 \,\, , \,\, G_2=T_2$

<a id='voltar1'></a>
[Exercício resolvido 1](#Link1) 

Mostre que as condições de equilíbrio na autarquia são:

$ \left\{ \begin{array}{l l} 
\cfrac{Y_2-G_2}{Y_1-G_1} = \cfrac{1+r}{1+ρ} \\
C_1 = Y_1 - G_1 , \\
C_2 = Y_2 - G_2 
\end{array}\right. $

♣

**Comentário**

(i) Um aumento no gasto público pode aumentar ou diminuir a taxa de juros, dependendo de quando ocorra. Um aumento no gasto futuro reduz a taxa de juros pois reduz a disponibilidade de recursos para o setor privado no futuro e estimula poupança para tranferir recursos para o futuro.

(ii) Caso a renda bruta seja igual nos dois períodos, variações de gastos iguais nos dois períodos não afetam a taxa de juros.

(iii) Esses resultados não são gerais. Eles dependem das hipóteses de renda exógena e impostos *lump-sum*. Mas eles são úteis para ilustrar como mudanças no gasto público afetam a disponibilidade de recursos para o setor privado e afetam as decisões de consumo e poupança.

(iv) O equilíbrio na pequena aberta é equivalente às condições de ótimos para o agente representativo que foram vistas no último capítulo.

**Exemplo 5**

Nada garante que a taxa de juros real na autarquia seja positiva. Se $Y_1>(1+ρ)Y_2$ a taxa de juros real será negativa. 

Suponha que: $(Y_1,Y_2, \rho)=(120, 100, 0.01)$.

A taxa de juros de equilíbrio será: -0.083, ou seja, -8,3%.

Na autarquia, agregadamente, não há como transferir recursos entre períodos. Se a renda corrente for (muito) superior à futura, haverá tal excesso de oferta de crédito que a taxa de juros cairá até que os indivíduos desistam de emprestar recursos. Essa taxa de juros (real) de equilíbrio pode ser negativa. Ou seja, os indivíduos estarão dispostos a pagar para conseguir transferir recursos para o futuro porque não há outra maneira de fazê-lo.

## Conclusão

(i) Na autarquia, a oferta agregada de recursos é totalmente inelástica e o equilíbrio é obtido exclusivamente por mudanças na taxa de juros (preço relativo) que deve variar até que os indivíduos escolham consumir em cada período exatamente os recursos disponíveis domesticamente. Na pequena economia aberta, a oferta é infinitamente elástica e o ajuste se dá por variações no volume de transações com o exterior (quantidades). Nesse caso, se todos os agentes são iguais, as quantidades de equilíbrio são um múltiplo das quantidades ótimas individuais.

(ii) Reduções (aumentos) na disponibilidade futura de recursos tendem a reduzir (aumentar) a taxa de juros na autarquia.

(iii) Os efeitos da política fiscal dependem não apenas de sua magnitude e sinal, mas também do seu perfil temporal, do seu *timing*. Modelos estáticos não são apropriados para discutir esse último aspecto.

### O pacote *NLsolve*

O equilíbrio na pequena economia aberta envolve resolver um sistema de equações que muda com o número de períodos da economia. À medida que o número de períodos aumenta isso rapidamente se torna impraticável. Por isso, em vez de calcular a solução analítica, utilizaremos um pacote para a solução de sistemas de sistemas de equações não-lineares. O pacote relevante é o [NLsolve](https://lectures.quantecon.org/jl/optimization_solver_packages.html#Systems-of-Equations-and-Least-Squares) que contém vários *solvers*, entre eles o **nlsolve** para solução de sistemas de equações não-lineares. [QuantEcon](https://lectures.quantecon.org/jl/optimization_solver_packages.html#Systems-of-Equations-and-Least-Squares) é uma  referência mais acessível para o *solver*.

Antes de passarmos ao código para resolver as equações, é preciso colocá-las no formato adequado, como explicado a seguir.

Os sistemas (A) e (B) abaixo são, obviamente, equivalentes. Entretanto, apenas (B) está no formato apropriado para utilização pelo *solver*. 

$ (A) \left\{ \begin{array}{l l}
-x^2+3xy=5 \\
3x=2xy^2-1
\end{array}\right.
\;\;\;\;\;
\Leftrightarrow
\;\;\;\;\;
(B) \left\{ \begin{array}{l l}
-x^2+3xy-5=0  \\
3x-2xy^2+1=0 \\
\end{array}\right. 
$

De forma geral, o *solver* assume que o sistema esteja no formato $F(x)=0$, onde $x$ é um vetor. Ao escrever as equações, digitamos, apenas, a parte $(F(x))$, omitindo o $(=0$); veja o código abaixo.

````julia
Pkg.add("NLsolve")
import NLsolve: nlsolve
Pkg.add("Plots")
import Plots: plot
````

````julia
function ex01(F, x)
    F[1] = -x[1]^2 + 3x[1]x[2] - 5
    F[2] = 3x[1] - 2x[1]x[2] + 1
end

result = nlsolve(ex01, [1.0 1.0])

x = result.zero
````

(i) Não confunda as variáveis no código do programa com as que aparecem no sistema original. Nesse exemplo: $x[1]=x\,,\,x[2]=y$. 

Cada variável do sistema original corresponde a uma posição no vetor das variáveis endógenas no código.

(ii) A função que contém o sistema $(\text{F(.)})$ tem tantos componentes (linhas) quantas forem as variáveis endógenas do sistema.

### Pequena economia aberta


O código abaixo computa o equilíbrio da pequena economia aberta e faz os gráficos das variáveis relevantes.



````julia
function pea(F, x)
    
    Ny = length(Y)
    Nr = length(r)
    
    if Ny != Nr                                 # verifica se os vetores têm o mesmo tamanho
        println("dimensões incompatíveis")      # envia mensagem de erro, se for o caso
    end
    
    for k = 1:Ny-1
        F[k] = x[k+1]/x[k] - (1+r[k+1])/(1+ρ)
    end
    
    F[Ny] = Y[1] - x[1] - x[Ny+1]
        
    
    for k = Ny+1:2Ny-2
        F[k] = Y[k-Ny+1] + (1+r[k-Ny+2])x[k] - x[k-Ny+1] - x[k+1]
    end
    
    F[2*Ny-1] = Y[2Ny-Ny] + (1+r[2Ny-Ny])x[2Ny-1] - x[2Ny-Ny]
    
    
end
````

````julia
function pea2()
    results = nlsolve(pea,  [ Y A] )
    x = results.zero
    
    Ny = length(Y)
    Nr = length(r)
    
    C = zeros(Ny,1)
    A2 = zeros(Ny,1)
    
    for k = 1:Ny
        C[k] = x[k]
    end
    
    for k = Ny+1:2Ny-1
        A2[k-Ny] = x[k]
    end
    
    
    time = range(1,stop=Ny)                # Períodos para o eixo horizontal do gráfico
    ρ2 =ρ*ones(Ny,1)
    
    p1 = plot(time, [Y' C], label=["Y" "C"])
    p2 = plot(time, A2, label=["A"])
    p3 = plot(time, [r' ρ2] , label=["r" "rho"])
    #pea2 = plot(p1,p2,p3, layout=(3,1))       # Saída da função: gráfico
    pea2 = plot(p1,p2,p3, layout=(3,1), fmt = :png, size = (800, 600) )     # Alternativa em formato exportável
end
````

````julia
ρ = 0.05; 
r  = [ρ 0.05 0.05 0.05 0.05]; 
 
Y = [100 110 120 110 100]; 
A = zeros(1,length(Y)-1)

pea2()
````

**Comentário**

(i) As variáveis exógenas são as rendas $(Y)$ e as taxas de juros $(r)$, que não precisam ser constantes.

(ii) O código utiliza duas funções: ```pea2()``` e ```pea(F,x)```, que é acionada por ```pea2()```.

(iii) A função ```pea(F,x)``` define as equações de equilíbrio para um número arbitrário de períodos. 

$\text{F}$ é um vetor com as equações de equilíbrio: as equações de Euler e as restrições orçamentárias. 

$\text{x}$ é um vetor que inclui os valores dos consumos $\text{(C)}$ e dos ativos $\text{(A)}$. 

Caso haja $\text{N}$ períodos, $\text{x}$ será um vetor com $\text{2N}$ posições. As primeiras $\text{N}$ para os consumos e as últimas $\text{N}$ para os ativos.

Por exemplo: $x[1] = C_1$ e $x[N+2] = A_1$. $x[N] = A_0 = 0$ é necessário por conta das dimensões dos vetores.

(iv) Para facilitar a compreensão do código dessa função, é útil escrever o código explicitamente para um determinado número de períodos. Por exemplo, para quatro períodos $(N = 4)$, o código na função **pea** é equivalente ao seguinte - com partes irrelevantes omitidas:

```
function pea(F, x)
    F[1] = x[2]/x[1] - (1+r[2])/(1+ρ)   
    F[2] = x[3]/x[2] - (1+r[3])/(1+ρ)
    F[3] = x[4]/x[3] - (1+r[4])/(1+ρ)
    
    F[4] = Y[1] - x[1] - x[5]
    
    F[5] = Y[2] + (1+r[2])x[5] - x[2] - x[6]
    F[6] = Y[3] + (1+r[3])x[6] - x[3] - x[7]
    
    F[7] = Y[4] + (1+r[4])x[7] - x[4]
end
```

<br/>Ou seja:

$ \left\{ \begin{array}{l l}
  \cfrac{C_2}{C_1} - \cfrac{1+r_1}{1+\rho} = 0
  \, \\
  \cfrac{C_3}{C_2} - \cfrac{1+r_2}{1+\rho} = 0
  \, \\
  \cfrac{C_4}{C_3} - \cfrac{1+r_3}{1+\rho} = 0
  \, \\
  \, \\
  Y_1 - C_1 - A_1 = 0
  \, \\
  \, \\
  Y_2 - C_2 - (1+r_1)A_1 - A_2 = 0
  \, \\
  Y_3 - C_3 - (1+r_2)A_2 - A_3 = 0
  \, \\
  \, \\
  Y_4 - C_4 - (1+r_3)A_3 = 0
  \, \\
  \end{array}\right.$


(v) A função ```pea2()``` não tem argumentos de entrada. A saída da função são três gráficos.

(vi) ```pea2()``` utiliza o **nlsolver** para resolver o sistema de equilíbrio. Para isso, é necessário especificar palpites iniciais para as variáveis endógenas $(C, A)$. Esses palpites são o segundo argumento do comando **nlsove**. Palpites razoáveis são: $C_i=Y_i \, , \, A_i=0$.

(vii) O sistema tem sete equações e sete endógenas: $ x[1]=C_1\,,\,x[2]=C_2\,,\,x[3]=C_3\,,\,x[4]=C_4\,,\,x[6]=A_1\,,\,x[7]=A_2\,,\,x[8]=A_3 $. Além disso, é acrescentada a variável endógena $x[5]=A_0=0$.

### Autarquia

No caso da autarquia, não há necessidade de utilizar o **nlsolve**. O consumos são iguais à renda em cada período e os juros saem diretamente da equação de Euler para cada par de períodos consecutivos.

````julia
Pkg.add("Plots")
import Plots: plot
````

````julia
function aut(Y,ρ)
    
    # Função para fazer o gráfico do juro de equilibrio na autarquia. 
    # Renda exógena. N períodos. Determinístico. 
    # Argumentos: vetor das rendas (Y) e escalar para a impaciência (ρ).
    
    
    # Número de períodos
    
    Ny = length(Y)
    

    # Juros
    
    r = zeros(Ny,1)
    r[1] = ρ
    
    for k = 1:Ny-1
        r[k+1] = (1+ρ)*(Y[k+1]/Y[k])-1      # Juro endógeno na autarquia. 
        end                                 # Euler para cada par de períodos consecutivos.           
    
    
    # Resultados
    
    ρ = ρ*ones(Ny,1)                       # Transforma um escalar em vetor para o gráfico
    time = range(1,stop=Ny)                # Períodos para o eixo horizontal do gráfico
    
    p1 = plot(time, [r ρ], label=["r" "rho"])   
    p2 = plot(time, Y, label ="Y=C")
    
    #aut = plot(p1,p2, layout=(2,1))       # Saída da função: gráfico.
    aut = plot(p1,p2, layout=(2,1), fmt = :png, size = (800, 600) )     # Alternativa em formato exportável
end
    

````

````julia
Y = [100, 110, 120, 110, 100]
ρ = 0.5

aut(Y,ρ)
````
