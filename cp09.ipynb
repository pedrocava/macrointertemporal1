## Introdução

Nos capítulos anteriores vimos como resolver modelos macroeconômicos com um número reduzido de períodos, dois ou, às vezes, três. Apesar de simples, esses modelos permitiram a análise de vários aspectos importantes das escolhas intertemporais dos agentes econômicos. Entretanto, para diversos problemas práticos, essa estrutura com poucos períodos é excessivamente rígida e precisamos de modelos em que um maior número de períodos permita maior flexibilidade na análise. Por exemplo, se quisermos analisar o desempenho da economia ao longo de cada um dos próximos seis meses, precisaremos de um modelo com vários períodos. 

Uma pergunta natural é: quantos? Até agora enfatizamos que os modelos devem ser mantidos tão simples quanto possível. Portanto, pareceria natural que para analisar a economia ao longo de um ano utilizássemos um modelo com doze períodos; para cinco anos, sessenta períodos. Essa abordagem, entretanto, não é prática. As contas ficam rapidamente intratáveis e precisariam ser refeitas a cada vez que o horizonte mudasse. A maneira prática é passar diretamente para um modelo com horizonte infinito. 

Mas essa é uma hipótese razoável? É claro que as pessoas não acham que viverão para sempre. Um horizonte de planejamento 
infinito é uma maneira simples de representar o fato de que as pessoas não sabem qual será o último período de suas vidas; e, portanto, não há um “último” período em que o indivíduo consome tudo que tem. (Essa modelagem acaba sendo mais simples que tratar a data da morte como uma variável aleatória – o que seria mais realista.) 

Mas um horizonte infinito não irá representar mal o comportamento das pessoas reais que vivem menos que isso? Na verdade, não, pois os valores futuros são descontados a taxas compostas, seja pela taxa de juros, seja pela de impaciência. Para valores razoáveis, um horizonte de cem ou de infinitos anos produz comportamentos muito parecidos.

Modelos com horizonte infinito são mais complexos que os de dois períodos que vimos até agora. Esses modelos raramente permitem uma solução analítica e precisam ser resolvidos graficamente ou numericamente, como veremos nos próximos capítulos.


## O modelo de crescimento em tempo discreto

Como primeiro exemplo, vamos estender o modelo de dois períodos com capital, mas sem lazer, ativos financeiros ou incerteza. Recapitulando, nesse caso, o problema do indivíduo é:

$ \left\{ \begin{array}{l l}	
\underset{X}{max} \, U(C_1,C_2)  = ln(C_1) + \cfrac{1}{1+\rho}ln(C_2)\\
    Y_1 = C_1 + I_1 \\
    Y_2 = C_2 + I_2   \\
    Y_1 = Z_1K_1^\alpha  \,\,\,\,\,\,  Y_2 = Z_2K_2^\alpha  \\
    K_2 = (1-\delta)K_1 + I_1 \,\,\,\,\,\,  K_3 = (1-\delta)K_2 + I_2  \\
    K_1,Z_1,Z_2: dados \,\,\,\,\,\, K_3 = K^F  \\
    \\
    X = (C_1,C_2,I_1,I_2,K_2,K_3,Y_1,Y_2 )   \\
    \end{array}\right. $

A primeira linha mostra o objetivo do indivíduo. A segunda mostra a restrição orçamentária do primeiro período e as restrições para cada estado da natureza possível no segundo período. A terceira faz o mesmo para as funções de produção. A quarta apresenta as equações de acumulação de capital, com uma taxa de depreciação δ. A quinta linha apresenta as probabilidades dos estados da natureza. A última apresenta a condição inicial e a final do problema.

Vamos, gradualmente, transformar esse problema para obter outro semelhante com horizonte infinito. 

Primeiro passo. (i) Definimos $β=\frac{1}{1+ρ}$.(ii) Substituir os valores das produções pelas suas expressões.

$ \left\{ \begin{array}{l l}	
    \underset{X}{max} \, U(C_1,C_2)  = ln(C_1) + \beta\ln(C_2)\\
    Z_1K_1^\alpha = C_1 + I_1 \\
    Y_2 = Z_2K_2^\alpha = C_2 + I_2   \\
    K_2 = (1-\delta)K_1 + I_1 \,\,\,\,\,\,  K_3 = (1-\delta)K_2 + I_2  \\
    K_1,Z_1,Z_2: dados \,\,\,\,\,\, K_3 = K^F  \\
    \\
    X = (C_1,C_2,I_1,I_2,K_2,K_3)   \\
    \end{array}\right. $

Segundo passo. Substituir os valores dos investimentos nas restrições orçamentárias nas equações de acumulação de capital.

$ \left\{ \begin{array}{l l}	
    \underset{X}{max} \, U(C_1,C_2)  = ln(C_1) + \beta\ln(C_2)\\
    K_2 = Z_1K_1^\alpha + (1-\delta)K_1 - C_1 \\
    K_3 = Z_2K_2^\alpha + (1-\delta)K_2  - C_2  \\
    K_1,Z_1,Z_2: dados \,\,\,\,\,\, K_3 = K^F  \\
    \\
    X = (C_1,C_2,K_2,K_3)   \\
    \end{array}\right. $

Terceiro passo. Vamos começar a contar o tempo em $t=0$ e não em $t=1$. Isso não tem maior importância, serve apenas para seguir a maioria dos textos.

$ \left\{ \begin{array}{l l}	
    \underset{ \{C_t,K_{t+1}\}}{max} \, U = \begin{equation}
              \sideset{}{_{t=0}^1}\sum\beta^tln(C_t)
            \end{equation} \\
    K_1 = Z_0K_0^\alpha + (1-\delta)K_0 - C_0 \\
    K_2 = Z_1K_1^\alpha + (1-\delta)K_1  - C_1  \\
    K_0,Z_0,Z_1: dados \,\,\,\,\,\, K_2 = K^F  \\
    \\
    X = (C_0,C_1,K_1,K_2)   \\
    \end{array}\right. $

Quarto passo. Note que as duas expressões para a acumulação de capital têm o mesmo formato. Isso nos permite escrever a equação de acumulação para um período *t* arbitrário. 

$ \left\{ \begin{array}{l l}	
    \underset{{\{C_t,K_{t+1}}\}}{max} \, U(C_t)  = \begin{equation}
              \sideset{}{_{t=0}^1}\sum\beta^tln(C_t)
            \end{equation} \\
    K_{t+}1 = Z_tK_t^\alpha + (1-\delta)K_t - C_t \\
    K_0,Z_0,Z_1: dados \,\,\,\,\,\, K_2 = K^F  \\
    \end{array}\right. $

Quinto passo. (i) O limite superior do somatório na função objetivo passa de um para infinito. (ii) Todos os valores da tecnologia serão dados exogenamente. (No próximo caopítulo, a trajetória da tecnologia será descrita por uma equação em diferenças.) (iii) No modelo de dois períodos, a condição final do problema é $K_2≥0$. Mas quando o horizonte é infinito não há um último período. Para simplificar a apresentação, vamos ignorar temporariamente essa condição – mas o que acontece no “final” dos tempos aparecerá nas condições de ótimo. A versão final do problema com horizonte infinito em tempo discreto é:

$ \left\{ \begin{array}{l l}	
    \underset{{\{ C_t,K_{t+1} }\}}{max} \, U  = \begin{equation}
              \sideset{}{_{t=0}^\infty}\sum\beta^tln(C_t)
            \end{equation} \\
    K_{t+1} = Z_tK_t^\alpha + (1-\delta)K_t - C_t \\
    K_0, Z_t: dados \,\,\,\,\,\,   \\
    \end{array}\right. $

Se você comparar a versão final do modelo com horizonte infinito com a versão inicial com dois períodos, notará que, embora a estrutura matemática seja mais complexa, a intuição econômica é a mesma. O problema do indivíduo consiste em escolher uma sequência de consumo e uma de capital (variáveis de controle) para maximizar uma soma ponderada de suas utilidades instantâneas (função objetivo) sujeito às restrições orçamentárias de cada período e ao estoque inicial de capital. Note que o indivíduos escolhe os consumos a partir de $t=0$ e o capital a partir de $t=1$, já que $K_0$ foi escolhido "ontem".

O que ganhamos com a maior complexidade? A possibilidade de tratar problemas econômicos em que dividir a vida em apenas dois períodos é uma simplificação excessiva. Agora, podemos falar de hoje, amanhã, depois de amanhã, depois de depois...


## A solução via Lagrange

Há mais de uma maneira de resolver o problema do indivíduo com horizonte infinito, visto na última seção. Vamos resolver formando o Lagrangeano do problema.

$   \underset{\{ C_t,K_{t+1},\lambda_t \}}{max} L  = 
              \sideset{}{_{t=0}^\infty}\sum { \left( \beta^tln(C_t) + 
              \lambda_t [ Z_tK_t^\alpha + (1-\delta)K_t - C_t - K_{t+1}] \right) }$

Em palavras. No período inicial $(t=0)$, o indivíduo planeja uma sequência de consumos e estoques de capital (e multiplicadores) ao longo da vida para maximizar sua expectativa de utilidade descontada ao longo da vida, com base nas informações inicialmente disponíveis.

No modelo com dois períodos, nós resolvemos o problema derivando em relação a cada variável de controle e a cada multiplicador. Aqui, entretanto, há uma infinidade de cada um deles. Na “força bruta” não conseguiremos resolver. A maneira de resolver o problema requer, em primeiro lugar, achar as condições de ótimo para um período “representativo” e, em seguida, resolver essas condições. Vejamos a primeira parte.

Considere a escolha do indivíduo em um período arbitrário *t* no futuro. Nesse período, ele deverá escolher: (i) o consumo daquele período e (ii) o estoque de capital do período seguinte. 

Para facilitar a visualização do problema, vamos escrever a parte relevante do Lagrangeano de maneira mais explícita.

$  \underset{ \{ C_t,K_{t+1},\lambda_t \}}{max} L  = 
    \,\,\, ... +\, \beta^tln(C_t) + \beta^{t+1}ln(C_{t+1}) \, + ... +\\
   \;\;\;\;\;\;\;\;\;\;\;\;\;\;\; + \lambda_t[Z_tK_t^\alpha + (1-\delta)K_t - C_t - K_{t+1}] \, +
    \lambda_{t+1}[ Z_{t+1}K_{t+1}^\alpha + (1-\delta)K_{t+1} - C_{t+1} - K_{t+2}] \, + ...
$
            
As condições de primeira ordem incluem:

$ \left\{ \begin{array}{l l}
\cfrac{\partial L}{\partial C_t} = \beta^t\cfrac{1}{C_t} - \lambda_t = 0 \\
\cfrac{\partial L}{\partial K_{t+1}} = \lambda_t - \lambda_{t+1}(\alpha Z_{t+1}K_{t+1}^{\alpha-1} 
    + 1-\delta) = 0  \\
\cfrac{\partial L}{\partial \lambda_t} = K_{t+1} - Z_tK_t^\alpha - (1-\delta)K_t + C_t = 0  \\
\end{array}\right.$

Ainda falta uma condição de ótimo, relacionada com o "período final". Mas, antes vamos manipular as duas primeiras condições acima e obter um resultado interessante.

Note que, como o período *t* é arbitrário:

$ β^t \cfrac{1}{C_t} = λ_t \,\,  \Rightarrow\,\,  β^{t+1} \cfrac{1}{C_{t+1}} = λ_{t+1} $

Substitua $λ_t$ e $λ_{t+1}$ na condição de ótimo para $K_{t+1}$:

$ β^t \cfrac{1}{C_t} - \left[ β^{t+1} \cfrac{1}{C_{t+1}} \left( \alpha Z_{t+1}K_{t+1}^{\alpha-1} + 1-\delta \right) \right] = 0  $

Simplificando:

$ \cfrac{1}{C_t} =  β \cfrac{1}{C_{t+1}} \left( \alpha Z_{t+1}K_{t+1}^{\alpha-1} + 1-\delta \right)    $

Mudando os índices de tempo, essa é a equação de Euler do modelo de dois períodos com capital. (Caso houvesse ativos financeiros, em equilíbrio a taxa de juros seria igual à produtividade marginal do capital menos a depreciação, a produtividade marginal líquida do capital.) A diferença importante é que, agora, temos uma equação em diferenças que precisa ser resolvida para acharmos toda a trajetória do consumo. Neste capítulo veremos uma solução gráfica e, no próximo, uma solução numérica.

Finalmente, precisamos de uma condição de ótimo para a escolha ótima no “final” do horizonte de planejamento. No modelo de dois períodos, a condição ótima era $K_3≥0$. Sem entrar em detalhes, a condição equivalente para o caso do horizonte infinito é:

$ \lim_{t\to \infty}{ \left[ \left(β^t \cfrac{1}{C_t} \right) K_{t+1} \right] } = 0 $

O termo dentro do parêntese interior é a utilidade marginal do consumo descontada pela impaciência. O termo dentro do colchete é o valor (em utilidade descontada) do estoque de capital. Finalmente, considerando o limite quando t→∞, temos o valor no “final” do horizonte de planejamento. Ou seja, não queremos deixar sobrar no final da vida nada que esperemos ser útil.

Resumindo, as condições de equilíbrio são:

$ \left\{ \begin{array}{l l}
\cfrac{1}{C_t} =  β \cfrac{1}{C_{t+1}} \left( \alpha Z_{t+1}K_{t+1}^{\alpha-1} + 1-\delta \right) \\
K_{t+1} = Z_tK_t^\alpha + (1-\delta)K_t - C_t   \\
K_0: dado \\
\lim_{t\to \infty}{ \left(β^t \cfrac{1}{C_t}K_{t+1} \right) } = 0  \\
\end{array}\right.$

Do ponto de vista econômico, a solução desse problema consiste em trajetórias para o estoque de capital e para o consumo que respeitem: (i) a escolha ótima dos indivíduos; (ii) a disponibilidade de recursos na economia; (iii) a evolução da tecnologia; (iv) o estoque inicial de capital; e (v) a condição final de ausência de desperdício. 

Do ponto de vista matemático, temos um sistema de duas equações em diferenças de primeira ordem, não lineares, para determinar duas variáveis endógenas – consumo e capital – com uma variável exógena – a tecnologia. As trajetórias devem respeitar as chamadas condições de contorno – o estoque inicial de capital e o seu valor “no infinito”.




## O diagrama de fases

A solução gráfica do problema requer a construção do que é conhecido como o “diagrama de fases” do problema. No nosso caso, o diagrama de fases é apenas um gráfico de $(C\, x\, K)$ no qual vamos representar: (i) o estado estacionário da economia e (ii)  as possíveis trajetórias conjuntas do consumo e do estoque de capital fora do estado estacionário. 

Para fazer o diagrama de fases é importante lembrar que, no estado estacionário: 

$C_t=C_{t+1}=C$, \, $K_t=K_{t+1}=K, \, Z_t=Z_{t+1}=Z$

Nossa estratégia é a seguinte. Primeiro, vamos determinar as combinações de consumo e capital que mantém o consumo constante – e vamos representar essas combinações no gráfico $(C\, x\, K)$. Em seguida, faremos a mesma coisa com as combinações que mantém o capital constante. Finalmente, veremos o que acontece com duas as variáveis quando elas não são constantes. 

A equação em diferenças que determina o movimento do consumo é:

$ \cfrac{1}{C_t} =  β \cfrac{1}{C_{t+1}} \left( \alpha Z_{t+1}K_{t+1}^{\alpha-1} + 1-\delta \right) $

Portanto:

$ C_t =  C_{t+1} ⇒ 1 = β \left(αZK_{t+1}^{α-1} +1-δ \right) $

Ou seja:

$ K_{t+1} = K^* = { \left[ \cfrac{1}{\alpha Z } \left( \cfrac{1}{\beta} - 1 + \delta \right) \right] } ^{\frac{1}{\alpha-1} } $

Essa expressão não depende do consumo e, portanto, é uma reta vertical $(\bar{C})$ no plano $(C\,x\,K)$; ver Figura 1. 



Quando $K_(t+1)=K^*$, o retorno por acumular uma unidade a mais de capital $\left( αZK_{t+1}^{α-1}+1-δ \right)$ – equivalente à taxa de juros – é igual à perda por esperar para consumir, a impaciência; nesse ponto, o consumo é constante, como vimos em outras versões da equação de Euler. À direita da reta vertical $K_{t+1}>K^*$, o retorno do investimento será menor que a impaciência (pois a produtividade marginal do capital $\left( αZK_{t+1}^{α-1} \right)$ é decrescente) e o consumo será decrescente. Verifique isso na expressão. De forma equivalente, à esquerda de $K^*$ o consumo será crescente. Isso é representado no gráfico pelas setas verticais.

Vejamos, agora, a condição para o estoque de capital ficar constante. A equação em diferenças que determina o movimento do capital é:

$ K_{t+1} = Z_tK_t^\alpha + (1-\delta)K_t - C_t $

$ K_t = K_{t+1} \,\, ⇒ \,\, C=ZK^α-δK $

Essa equação indica que o estoque de capital ficará constante quando o consumo for igual à produção líquida. A equação é representada por uma curva que sai da origem, é inicialmente crescente, atinge um máximo e decresce; curva (k ̇ ). Em pontos acima da curva, o consumo é superior à produção líquida, o estoque de capital diminui e as setas horizontais apontam para a direita. De forma equivalente para pontos abaixo da curva.

Resumindo, o estado estacionário é dado por:

$ \left\{ \begin{array}{l l}
1 = β \left(αZK_{t+1}^{α-1} +1-δ \right) \\
C=ZK^α-δK \\
\end{array}\right.$

Antes de prosseguir com a análise do gráfico, é preciso considerar a segunda simplificação mencionada acima. Na discussão acima, nós implicitamente consideramos que cada período é bastante curto e que $K_t \approx K_{t+1}$. Sem essa aproximação a curva que determina $C_t = C_{t+1}$ não seria vertical, mas positivamente inclinada. Se considerarmos o limite quando os períodos se tornam cada vez mais curtos, teremos um modelo em tempo contínuo e a curva será vertical sem aproximações. Entretanto, para nossos objetivos, não é necessária uma discussão mais detalhada.

A Figura 2 mostra uma trajetória da economia a partir de uma posição inicial arbitrária. No ponto B: (i) a taxa de juros é superior à impaciência, o consumo aumenta e a economia se move para cima; (ii) o consumo é inferior à renda, o estoque de capital aumenta e a economia se move para a direita. O resultado é que a economia se desloca, ao longo tempo, na direção nordeste. Neste exemplo, a economia atinge a curva (k ̇ ) antes da (c ̇ ) e, a partir daí segure na direção noroeste rumo a D. Ao atingir o eixo vertical, o estoque capital acaba, não há mais renda nem consumo e a economia vai para a origem, que também é um estado estacionário. (Note que no ponto C, o movimento da economia é exclusivamente vertical para cima.)



A Figura 3 mostra várias trajetórias possíveis. Vemos que há apenas duas curvas que levam ao estado estacionário. Além disso, é possível mostrar que duas trajetórias jamais se cruzam e que por cada ponto do gráfico passa uma curva. A Figura 3 indica que para cada nível de capital existe apenas um nível de consumo que coloca a economia em uma trajetória que leva ao estado estacionário, a chamada trajetória convergente ou  saddle-path. Se, para um nível de capital inicial, consumo inicial for muito grande ou muito pequeno, a economia não irá para o estado estacionário.







A discussão até agora ignorou a condição inicial $(K_0:\,dado)$ e a condição final do sistema $\left(\lim_{t\to \infty}{ \left(β^t \frac{1}{C_t}K_{t+1} \right) } = 0  \right)$. Vejamos o que elas nos dizem.

Como o nome sugere, a condição inicial indica a posição de partida da economia. Essa condição especifica apenas o estoque de capital inicial; o consumo inicial não é um dado do problema, mas uma escolha do indivíduo. Qualquer que seja o consumo inicial escolhido pelo indivíduo, ele deverá estar sobre uma reta vertical que passe por $K_0$. Caso contrário, ou o indivíduo estaria destruindo capital inutilmente ou criando capital a partir do nada.

Uma vez que o indivíduo tenha escolhido o consumo inicial, a economia segue a trajetória descrita pela curva correspondente. Uma trajetória que comece acima do saddle-path não pode ser uma escolha ótima, pois, a longo prazo, o estoque de capital irá para zero, o indivíduo não terá mais recursos e morrerá de fome. A utilidade marginal do capital será tão grande que a condição final não será respeitada. Da mesma forma, uma trajetória que comece abaixo do saddle-path tampouco será ótima, pois o consumo também irá para zero. O gráfico indica que apenas as trajetórias que levem ao estado estacionário podem ser ótimas e isso determina o consumo inicial ótimo.

O comportamento ótimo pode ser resumido em dois passos: (i) escolher o consumo inicial ótimo e (ii) escolher os consumos subsequentes respeitando  equação de Euler. O consumo inicial ótimo é aquele que, seguindo a equação de Euler no futuro, exatamente esgota a restrição orçamentária intertemporal (incluindo a condição final). 


## Exemplos

Exemplo 1: Suponha que a economia esteja inicialmente no estado estacionário e que uma catástrofe natural destrua parte do estoque de capital. O estoque de capital passa de $K_0$ para $K_1$ na Figura 4 – note que essa redução não foi escolhida pelos agentes. A escolha ótima coloca a economia no ponto *B*. Nesse ponto, os indivíduos consomem menos que a renda líquida, e o estoque de capital aumenta. Além disso, inicialmente o retorno do investimento está acima da impaciência e os indivíduos desejam consumo crescente. 



O resultado é bastante intuitivo. A destruição do estoque de capital não alterou as condições estruturais da economia: a função de produção, as preferência e a tecnologia continuam as mesmas. Portanto, o estado estacionário continua o mesmo. Para poderem recuperar seu estoque de capital, os indivíduos reduzem seu consumo inicial abaixo da nova e menor renda líquida e, à medida que vão ficando mais ricos, passam a consumir mais até voltarem ao estado estacionário original. 


Comentário

(i) O consumo pode dar saltos de um período para o seguinte. Mas o estoque de capital não pode: não vamos destruir capital e não podemos criá-lo do nada.

(ii) Nesse modelo, embora o consumo possa dar saltos em qualquer período a escolha ótima requer os eventuais saltos ocorram apenas no período inicial. Um salto planejado ao longo da trajetória futura violaria a equação de Euler e não seria ótimo. Um salto do consumo no período inicial responde a um choque inesperado, não foi planejado e não tem como ser evitado. 

(iii) O problema é resolvido de trás para frente. Primeiro: determine o estado estacionário (posição final). Segundo: determine a (única) trajetória convergente que leva ao estado estacionário. Terceiro: Determine a (única) posição inicial da variável de controle que coloca a economia sobre a trajetória convergente.	♠




### Política fiscal simplificada

Vamos considerar algumas variações no gasto público. Para concentrar em um caso bastante simples, suponha que: (i) o governo mantenha o orçamento equilibrado a cada período $(G_t=T_t)$; (ii) o gasto público não afete nem a utilidade dos agentes nem a produção da economia; e (iii) que os impostos sejam *lump-sum*. Nesse caso, o problema do indivíduo passa a ser:

$ \left\{ \begin{array}{l l}	
    \underset{{{ C_t,K_{t+1} }}}{max} U  = \begin{equation}
              \sideset{}{_{t=0}^\infty}\sum\beta^tln(C_t)
            \end{equation} \\
    K_{t+1} = Z_tK_t^\alpha + (1-\delta)K_t - C_t - T_t \\
    K_0: dado \,\,\,\,\,\,   \\
    \lim_{t\to \infty}{ \left(β^t \cfrac{1}{C_t}K_{t+1} \right) } = 0  \\
    \end{array}\right. $

É fácil ver que o equilíbrio da economia será dado por:

$ \left\{ \begin{array}{l l}
\cfrac{1}{C_t} =  β \cfrac{1}{C_{t+1}} \left( \alpha Z_{t+1}K_{t+1}^{\alpha-1} + 1-\delta \right) \\
K_{t+1} = Z_tK_t^\alpha + (1-\delta)K_t - C_t  -G_t \\
K_0: dado \\
\lim_{t\to \infty}{ \left(β^t \cfrac{1}{C_t}K_{t+1} \right) } = 0  \\
\end{array}\right.$

E as condições de estacionariedade das variáveis são:

$ \left\{ \begin{array}{l l}
C = ZK^α - δK -G \\
1 = β \left(αZK^{α-1} +1-δ \right) \\
\end{array}\right.$



A única modificação em relação ao caso sem governo é a presença do gasto público na equação de acumulação. Essa simplicidade depende das três hipóteses acima sobre a política fiscal.


Exemplo 2: Considere um aumento inesperado e permanente no gasto público.

As condições de estacionariedade mostram que a curva $(\bar{C})$ não se desloca e que a curva $(\bar{K})$ se desloca paralelamente para baixo: o consumo privado é integralmente deslocado pelo gasto do governo.  Se a economia está inicialmente em equilíbrio no ponto A, o consumo é reduzido para B e a economia passa instantaneamente para seu novo estado estacionário, sem nenhuma dinâmica de transição – ver Figura 5.



É claro que do ponto de vista do bem-estar o novo estado estacionário é inferior ao original, ou seja, o aumento no gasto público reduziu o bem-estar dos agentes. Embora isso seja verdade nesse modelo simplificado, é preciso tomar algum cuidado com essa conclusão. 	♣

Exemplo 3: Aumento antecipado e permanente no gasto público.  

Quando ocorrer, o aumento no gasto do governo deslocará a curva (K ̇ ) paralelamente para baixo.  O estado estacionário passará de A para D – ver Figura 6.  



Para descobrir a trajetória de transição ótima, observe que, enquanto o aumento anunciado do gasto do governo não tiver ocorrido, as curvas relevantes são (K ̇ ) e (C ̇ ).  Quando o gasto for efetivamente aumentado, as curvas relevantes passam a ser  (K ̇^' ) e (C ̇ ). No momento em que o gasto do governo for efetivamente alterado, a economia deve estar em algum ponto sobre o saddle path que leva a D. Se não estiver, terá que haver um salto no nível de consumo.  Mas isso implica uma descontinuidade planejada na utilidade marginal do consumo – o que viola as condições para a maximização de utilidade.  Portanto, no instante em que o aumento no gasto é anunciado, os agentes devem ajustar seu consumo de forma que a economia atinja o saddle path que leva a D no instante em que o gasto for efetivamente aumentado.  

A Figura 6 mostra que, inicialmente, o consumo cai de A para B.  No ponto B, o campo de força da economia aponta na direção leste e, imediatamente à direita de B, na direção sudeste.  (O consumo do governo ainda não aumentou e as curvas relevantes continuam sendo (K ̇ ) e (C ̇ ).) A partir de B, a economia desloca se em direção ao ponto sobre o saddle path que leva a D. Durante esta etapa, o consumo cai e o estoque de capital aumenta.  O aumento no estoque de capital implica na queda da sua produtividade marginal e, portanto, no retorno do capital.  A trajetória de B para C é uma etapa de aumento de poupança para fazer face ao aumento esperado no gasto do governo: os agentes reduzem seu consumo/aumentam sua poupança.  Lembre que acumular capital é a única forma dessa economia fechada poupar agregadamente. A economia atinge o ponto C exatamente no instante em que o governo aumenta efetivamente os gastos.  Nesse instante, as novas curvas relevantes são (K ̇ ) e (C ̇ ), a economia está sobre o novo saddle path e se dirigirá, sem novos saltos, para D. 

Supusemos que os agentes consigam calcular exatamente qual o valor do consumo inicial necessário para atingirem o estado estacionário final sem saltos antecipados no consumo. É claro que as pessoas não têm esse nível de precisão. O que acontece quando elas percebem que erraram seus cálculos e caminham para uma situação insustentável? Quando percebem isso, as pessoas refazem seus planos e ajustam seu consumo de forma a passar a uma trajetória que julguem sustentável. As trajetórias reais terão uma série de ajustes não planejados em torno da trajetória ideal.
Como sempre, um modelo é uma simplificação da realidade que visa capturar as características mais importantes do comportamento da economia. Nesse exemplo, o que é qualitativamente importante é que as pessoas começam a poupar antes do governo aumentar o gasto e, quando os impostos efetivamente aumentam, passam a despoupar. Sabendo que ficarão mais pobres no futuro, as pessoas transferem recursos do período de vacas gordas para o de vacas magras.	♣


## Conclusão

Este capítulo mostrou como estender nosso modelo básico para um número infinito de períodos e como fazer uma análise gráfica do equilíbrio da economia. Lembre que os gráficos estão baseados em uma aproximação supondo que o tamanho dos períodos seja pequeno; em tempo contínuo, os gráficos estariam exatos. É claro que os gráficos fornecem apenas soluções qualitativas. Veremos a seguir como resolver esse modelo numericamente para obter resultados mais precisos.


## Resumo

Problema do indivíduo

$ \left\{ \begin{array}{l l}	
    \underset{{\{ C_t,K_{t+1} }\}}{max} \, U  = \begin{equation}
              \sideset{}{_{t=0}^\infty}\sum\beta^tln(C_t)
            \end{equation} \\
    K_{t+1} = Z_tK_t^\alpha + (1-\delta)K_t - C_t \\
    K_0, Z_t: dados \,\,\,\,\,\,   \\
    \end{array}\right. $
    
Equilíbrio

$ \left\{ \begin{array}{l l}
\cfrac{1}{C_t} =  β \cfrac{1}{C_{t+1}} \left( \alpha Z_{t+1}K_{t+1}^{\alpha-1} + 1-\delta \right) \\
K_{t+1} = Z_tK_t^\alpha + (1-\delta)K_t - C_t   \\
K_0: dado \\
\lim_{t\to \infty}{ \left(β^t \cfrac{1}{C_t}K_{t+1} \right) } = 0  \\
\end{array}\right.$
