## Introdução

Neste capítulo, vamos calcular o equilíbrio na economia com investimento.

Como vimos anteriormente, o equilíbrio em uma pequena economia aberta gera as mesmas condições que as condições de ótimo para o indivíduo, pois a taxa de juros é exógena e os indivíduos têm acesso a um mercado de capitais ideal, sem restrições. Essa é uma hipótese crítica. Sem acesso a um mercado de capitais ideal, alguns resultados importantes da última seção não valem. Uma autarquia não pode se financiar com o resto do mundo e, portanto, permite analisar como os resultados anteriores mudam quando, agregadamente, não há acesso ao mercado de capitais. 


## Equilíbrio na pequena economia aberta

Sabemos que na pequena economia aberta as condição de equilíbrio são um múltiplo das condições de ótimo do agente representativo. Portanto, as condições de equilíbrio incluem:

$ \left\{ \begin{array}{l l}
    r=\alpha Z_2K_2^{\alpha-1} - \delta  \\
    Z_1K_1^α  + \cfrac{Z_2K_2^α}{1+r} = C_1 + K_2 - (1-\delta)K_1 +
    \cfrac{C_2 + K^F - (1-\delta)K_2}{1+r}\\
    \cfrac{C_2}{C_1}=\cfrac{1+r}{1+\rho}   \\
  \end{array}\right. $

Como vimos, esse sistema pode ser resolvido de forma recursiva. Dada a taxa de juros internacional (exógena) a primeira equação determina o estoque ótimo de capital do segundo período $(K_2^*)$. Dado o capital ótimo, as duas outras equações determinam conjuntamente os consumos ótimos $(C_1^*,C_2^*)$. 

**Comentário**

(i) Uma variação na tecnologia do primeiro período não afeta o investimento no primeiro período. O investimento no primeiro período produz capital que só entra em operação no segundo período. Portanto, a produtividade marginal relevante para a decisão de investimento hoje é a produtividade do capital amanhã, que não é afetada pelo choque de tecnologia hoje. Por outro lado, um aumento na tecnologia do segundo período eleva o investimento corrente. 

Note que estamos supondo que as variações na tecnologia são independentes. Se a mudança fosse persistente, a variação na tecnologia corrente afetaria o investimento apenas indiretamente via seu efeito na tecnologia futura. Esse resultado depende do acesso irrestrito ao mercado de capitais. Veremos adiante que ele não será verdadeiro em uma autarquia. 

(ii) O estoque de capital ótimo é obtido resolvendo a primeira condição de equilíbrio.

$ r=\alpha Z_2K_2^{\alpha-1} - \delta \,\,\,
\Rightarrow \,\,\,
 K_2^* = \left( \cfrac{\alpha Z_2}{r+\delta}\right)^\cfrac{1}{1-\alpha}  $

A expressão mostra que um aumento na taxa de juros ou na taxa de depreciação reduz o investimento, pois eleva o custo do capital e requer uma elevação na produtividade marginal do capital para ser rentável; lembre que a produtividade marginal do capital é decrescente. Um aumento na tecnologia tem o efeito oposto. 

Lembrando que $I_2^*=K_2^*-K_1$, vemos que cada unidade adicional de capital inicial reduz o investimento necessário para atingir esse nível em uma unidade.

(iii) Tendo calculado $(K_2^*, \, C_1^*, \, C_2^* )$, as demais endógenas são obtidas resolvendo:

$ \left\{ \begin{array}{l l}
    Y_1 = Z_1K_1^\alpha  \\
    Y_2 = Z_2K_2^{*\alpha}  \\
    I_1 = K_2^* - (1-\delta)K_1  \\
    I_2 = K^F - (1-\delta)K_2^* \\ 
    A_1 = Y_1 - C_1^* - I_1 \\
 \end{array}\right. $

## Equilíbrio na autarquia

Em autarquia, os ativos financeiros agregados são zero e as restrições orçamentárias agregadas implicam:

$ \left\{ \begin{array}{l l}
Z_1 K_1^α = C_1 + K_2-(1-\delta)K_1 \\
Z_2 K_2^α = C_2 + (1-\delta)K_2-K^F   \\
\end{array}\right. $

A primeira equação mostra que a produção agregada do primeiro período é dividida entre o consumo e o investimento do período. A segunda equação mostra que o consumo do segundo período é igual à produção mais o desinvestimento. 

Para obter as condições de equilíbrio, precisamos combinar as escolhas ótimas com as restrições agregadas de recursos. O <a id='voltar1'></a> [Exercício resolvido 1](#Link1) mostra que as condições de equilíbrio incluem:

 $\left\{ \begin{array}{l l}
r = \alpha Z_2K_2^{\alpha-1} - \delta  \\ 
(1+\rho)[Z_2 K_2^α + (1-\delta)K_2 - K^F] =
(1+r)[Z_1K_1^α + (1-\delta)K_1 - K_2]  \\
\end{array}\right. $

O sistema não é mais recursivo: ambas as equações envolvem as endógenas $(K_2,r)$. Portanto, a escolha do nível ótimo de capital para o segundo período dependerá do valor do capital inicial e da impaciência. Indivíduos mais impacientes ou inicialmente mais pobres tenderão a investir menos e a ter um estoque de capital menor, abrindo mão de oportunidades lucrativas de investimento para “não passarem fome no presente”. 

A próxima seção apresenta exemplos numéricos para ilustrar o funcionamento dessa economia simplificada.


### Autarquia sem ativos financeiros

Na autarquia, como os ativos financeiros agregados somam zero, podemos escrever o modelo sem os ativos financeiros.

$ \left\{ \begin{array}{l l}	
\underset{X}{max} U(C_1,C_2)  = ln(C_1) + \cfrac{ln(C_2)}{1+\rho}\\
    Y_1 = C_1 + I_1 \\
    Y_2 = C_2 + I_2   \\
    Y_1 = Z_1K_1^\alpha  \,\,\,\,\,\,  Y_2 = Z_2K_2^\alpha  \\
    K_2 = (1-\delta)K_1 + I_1 \,\,\,\,\,\,  K_3 = (1-\delta)K_2 + I_2  \\
    K_3 = K^F  \\
    \\
    X = (C_1,C_2,I_1,I_2,K_2,K_3,Y_1,Y_2 )   \\
    \end{array}\right. $
    
O equilíbrio será o mesmo que no modelo com ativos financeiros. Podemos até, em um abuso de linguagem, falar da "taxa de juros" para nos referirmos à produtividade marginal líquida do capital $ \left(\alpha Z_2K_2^{\alpha-1} + 1 - \delta \right)$. Essa formulação será utilizada algumas vezes mais adiante.

## Exemplos

Vejamos alguns exemplos numéricos da reação da economia a choques exógenos. 

**Exemplo 1** 

Aumento na tecnologia do segundo período em uma autarquia. Suponha que inicialmente:

$ K_1=100; \,\, Z_1 = Z_2 = 1; \,\, α = 0.5; 
  \delta = 0.0; \,\, ρ = 0.05; \,\, K^F = K_2  $

O equilíbrio em autarquia será:

$ Z_1 =  1; \,\, Z_2 =  1; \,\, K_1 =  100; \,\, K_2 =  100; K_3 =  100; \\
Y_1 =  10; \,\, Y_2 =  10; \,\, r =  0.050; \,\, C_1 =  10; \,\, C_2 =  10 $

Suponha que a tecnologia do segundo período aumente: $Z_2 = 1.01$. O novo equilíbrio será:

$Z_1 =  1; \,\, Z_2 =  1.0100; \,\, K_1 =  100; \,\, K_2 =  99.954; \,\, K_3 = 99.954; \\Y_1 =  10; \,\, Y_2 =  10.098; \,\, r =  0.050512; \,\,
C_1 =  10.046; \,\, C_2 =  10.051. $

O aumento na tecnologia do segundo período tende a elevar a renda desse período. Os indivíduos tentam elevar seu consumo inicial antecipando esse aumento na renda. Os juros sobem, ficando acima da impaciência. O estoque de capital diminui e o consumo é crescente. 

Note a diferença em relação à pequena economia aberta em que o aumento na tecnologia $(Z_2)$ eleva o estoque de capital $(K_2)$.

## Conclusão

Este capítulo discutiu o problema de alocação intertemporal de recursos de um consumidor que tem acesso a investimentos financeiros e em capital físico. Isso significa que a renda (do segundo período) passa a ser endógena. A renda também seria endógena caso o indivíduo pudesse escolher o número de horas a trabalhar, como vermos em um próximo capítulo.

Vimos que a condição para a escolha do perfil ótimo de consumo ao longo do tempo não muda e a equação de Euler continua valendo. Esse é um resultado bastante robusto que aparecerá nos diversos modelos que veremos.

Em segundo lugar, vimos que, na presença de um mercado de capitais ideal, o consumidor pode separar sua decisão de investimento da sua decisão de consumo. O indivíduo investirá até que a produtividade marginal do capital se iguale ao custo do capital – a taxa de juros –, financiando suas despesas no mercado de capitais. Esse resultado não vale se a hipótese de um mercado de capitais ideal não for uma boa aproximação para o problema que estivermos estudando. Em termos agregados, o resultado não vale em uma autarquia.

**Problema individual**

$ \left\{ \begin{array}{l l}	
\underset{X}{max} U(C_1,C_2)  = ln(C_1) + \cfrac{ln(C_2)}{1+\rho}\\
    Y_1 = A_1 + C_1 + I_1 \\
    Y_2 + (1+r)A_1 = C_2 + I_2  + A_2  \\
    Y_1 = Z_1K_1^\alpha  \,\,\,\,\,\,  Y_2 = Z_2K_2^\alpha  \\
    K_2 = (1-\delta)K_1 + I_1 \,\,\,\,\,\,  K_3 = (1-\delta)K_2 + I_2  \\
    A_2= 0 \, , \,\,\, K_3 = K^F  \\
    \\
    X = (C_1,C_2,I_1,I_2,A_1,A_2,K_2,K_3,Y_1,Y_2 )   \\
    \end{array}\right. $

**Condições de ótimo**

$ \left\{ \begin{array}{l l}
r = \alpha Z_2K_2^{\alpha-1} - \delta  \\
Z_1.K_1^α - \delta K_1 + \cfrac{1}{1+r}(Z_2.K_2^α -\delta K_2) = C_1 + (K_2 - K_1) +
\cfrac{1}{1+r}\left[ C_2 + (K^F - K_2) \right]  \\
\cfrac{C_2}{C_1}=\cfrac{1+r}{1+\rho}   \\
\end{array}\right. $


**Equilíbrio em autarquia**

$ \left\{ \begin{array}{l l}
r = \alpha Z_2K_2^{\alpha-1} - \delta  \\ 
(1+\rho)\left[Z_2 K_2^α + (1-\delta)K_2 - K^F \right] =
(1+r)\left[Z_1K_1^α - K_2 + (1-\delta)K_1 \right]  \\
\end{array}\right. $

Antes de passarmos ao código para resolver as equações, é preciso colocá-las no formato adequado, como explicado a seguir.

Os sistemas (A) e (B) abaixo são, obviamente, equivalentes. Entretanto, apenas (B) está no formato apropriado para utilização pelo *solver*. 

$ (A) \left\{ \begin{array}{l l}
-x^2+3xy=6 \\
3x^2=2xy^2-4
\end{array}\right.
\;\;\;\;\;
\Rightarrow
\;\;\;\;\;
(B) \left\{ \begin{array}{l l}
-x^2+3xy-6=0  \\
3x^2-2xy^2+4=0 \\
\end{array}\right. 
$

De forma geral, o *solver* assume que o sistema esteja no formato $F(x)=0$, onde $x$ é um vetor. Ao escrever as equações, digitamos, apenas, a parte $(F(x))$, omitindo o $(=0$); veja o código abaixo.

As próximas células apresentam o código.

```julia
# Esta célula  carrega o pacote NLsolve e importa o solver nlsolve. 
# Ela só precisa ser executada uma vez.

Pkg.add("NLsolve")
import NLsolve: nlsolve
```

O programa na próxima célula calcula o equilíbrio em autarquia em duas etapas. (i) Cálculo do estoque de capital e da taxa de juros de equilíbrio. (ii) Cálculo das demais variáveis econômicas.

As equações que definiem o equilíbrio são:

$ \left\{ \begin{array}{l l}
r = \alpha Z_2K_2^{\alpha-1} - \delta  \\ 
(1+\rho)\left[Z_2 K_2^α + (1-\delta)K_2 - K^F \right] =
(1+r)\left[Z_1K_1^α - K_2 + (1-\delta)K_1 \right]  \\
\end{array}\right. $

Uma vez resolvido esse bloco (simultâneo), as demais variáveis endógenas são dadas pela aplicação direta das fórmulas; por exemplo, a função de produção determina o produto dado o estoque de capital de equilíbrio.

As variáveis endógenas serão armazenadas no vetor $x$, que terá duas posições: $x[1]=r$ e $x[2]=K_2$.

```julia
# Utilidade CES. Log: v=1.

# Exógenas e parâmetros

Z1 = 1.01; Z2 = 1.01;  α = 0.4; δ = 0.0; ρ = 0.05; v  = 1; 

K1 = (α*Z1/(ρ+δ))^(1/(1-α))   # Calculado a partir de um modelo estático


# Este bloco define o sistema de equações a ser resolvido
# x é um vetor com, neste caso, duas posições.
# A primeira equação é a igualdade entre juros (x[2]) e produtividade marginal do capital (x[1]).
# A segunda equação é a equação de Euler.

f(x) = [ Z2*α*x[1]^(α-1) - δ - x[2]   
    
        (1+ρ)*(Z2*x[1]^α+(1-δ)*x[1]-x[1])^v - (1+x[2])*(Z1*K1^α+(1-δ)*K1-x[1])^v ]  # returns an array


# Este bloco utiliza o solver nlsolve
# f é o sistema definido acima.
# [100; 0.05] é o palpite inicial de solução para o capital e os juros.
# Em conjunto, as duas linhas de código armazenam os resultados no vetor x.

results = nlsolve(f, [ 100; 0.05])
x = results.zero


# Resultados

# Capital e juros
K2 = x[1]
r = x[2]

# Demais endógenas
Y1 = Z1*K1^α
Y2 = Z2*K2^α
C1 = Y1 + (1-δ)*K1 - K2
C2 = Y2 + (1-δ)*K2 - K2

# Exibição dos resultados
[K1 K2; r r; Y1 Y2; C1 C2]
```
