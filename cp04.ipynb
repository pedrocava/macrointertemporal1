## Introdução


A discussão anterior nos levou à seguinte versão simplificada do problema do consumidor:

$ \left\{ \begin{array}{l l}	
\underset{(c_1,c_2,a_1)}{max} U(c_1,c_2)  = ln(c_1) + \cfrac{ln(c_2)}{1+\rho}\\
     y_1 = c_1 + a_1 \\
     y_2 + (1+r)a_1 = c_2  \\
    \end{array}\right. $

Recapitulando. Queremos analisar as escolhas ótimas de consumo e poupança de um indivíduo que: (i) vive dois períodos; (ii) recebe uma dotação exógena de bens perecíveis $(y_1,y_2)$; (iii) obtém satisfação apenas do consumo desses bens $(c_1,c_2)$; (iv) tem acesso irrestrito a um ativo financeiro $(a_1)$ a uma taxa de juros $(r)$ fixa; (v) tem o benefício da previsão perfeita, ou seja, não enfrenta incerteza quanto ao futuro.

Neste capítulo, queremos resolver o problema acima. Isso significa calcular a funções que determinam o consumo ótimo em cada período e a poupança (variáveis endógenas) em função das rendas e da taxa de juros (variáveis exógenas) e da taxa de impaciência (parâmetro). 

Note que, embora o indivíduo tenha três variáveis de escolha, ele está sujeito a duas restrições e, portanto, tem apenas um grau de liberdade. Suponha que o indivíduo escolha o consumo do primeiro período. Como a renda do primeiro período é exógena, isso determina a poupança. No segundo período, a renda e os ativos financeiros estarão dados e o indivíduo consumirá todos os seus recursos, sem nenhuma nova escolha.

## Solução

Vamos resolver o problema do consumidor via [*Lagrange*](https://www.khanacademy.org/math/multivariable-calculus/applications-of-multivariable-derivatives/lagrange-multipliers-and-constrained-optimization/v/the-lagrangian). <a id='voltar1'></a> O [Exemplo 1](#Link1) apresenta formas alternativas de resolver o problema.

 $   \underset{(c_1,c_2,a_1,\lambda_1,\lambda_2)}{max} L = ln(c_1) + \cfrac{ln(c_2)}{1+\rho} + 
        \lambda_1(y_1 - c_1 - a_1) + \lambda_2[y_2 + (1+r)a_1 - c_2]  \\ $
        
As condições de primeira ordem para o ótimo são:

$ 
\left\{ \begin{array}{l l}
(i)\,\, \cfrac{\partial L}{\partial c_1} = \cfrac{1}{c_1} - \lambda_1 = 0 \,\, \Rightarrow\,\, \cfrac{1}{c_1} = \lambda_1  \\
(ii)\,\, \cfrac{\partial L}{\partial c_2} = \cfrac{1}{1+\rho} \cfrac{1}{c_2} - \lambda_2 = 0 \,\,  \Rightarrow\,\, \cfrac{1}{1+\rho} \cfrac{1}{c_2} = \lambda_2\\
(iii)\,\, \cfrac{\partial L}{\partial a_1} = - \lambda_1 + (1+r)\lambda_2 = 0 \,\,  \Rightarrow\,\, \lambda_1 = (1+r)\lambda_2\\
(iv)\,\, \cfrac{\partial L}{\partial \lambda_1} = y_1 - c_1 - a_1 = 0 \,\,  \Rightarrow\,\, y_1 = c_1 + a_1\\
(v)\,\, \cfrac{\partial L}{\partial \lambda_2} = y_2 + (1+r)a_1 - c_2 = 0 \,\, \Rightarrow\,\, y_2 + (1+r)a_1 = c_2\\
\end{array}\right.
$

As condições de ótimo formam um sistema com cinco equações que determina as cinco variáveis endógenas do problema: $ c_1,c_2,a_1,\lambda_1,\lambda_2 $. Antes de resolvermos esse sistema, vamos interpretar as condições acima.

**Comentário**

A interpretação das condições de ótimo é a seguinte. (Lembre que o multiplicador representa o *preço sombra* dos recursos exógenos. <a id='voltar2'></a> O [Exemplo 2](#Link2) ilustra essa interpretação.)

(i) O lado esquerdo da equação $(i)$ indica a utilidade marginal do consumo do primeiro período: $ \cfrac{\partial U(C_1,C_2)}{\partial C_1} = \cfrac{1}{C_1} $. E o lado direito $(\lambda_1)$ indica, como vimos, a utilidade marginal do recurso exógeno associado àquela restrição, no caso, a renda do primeiro período $(Y_1)$. Em cada período, o indivíduo tem a opção de consumir seus recursos imediatamente ou não consumí-los e guardá-los para consumir mais tarde. A equação (i) indica que, no ponto de ótimo, o ganho marginal de consumir imediatamente deve ser igual ao de não consumir imediatamente. Se esses ganhos marginais não fossem iguais, o indivíduo poderia aumentar sua satisfação total realocando seus recursos.

(ii) A interpretação dessa condição é semelhante à da anterior. A única diferença é que se refere a valores do segundo período que precisam ser trazidos a valor presente para o primeiro período, que é o momento da tomada de decisão. Dessa forma, o lado esquerdo da equação indica a utilidade marginal do consumo do segundo período trazida a valor presente pela taxa de impaciência. E o lado direito indica a utilidade marginal dos recursos futuros trazida a valor presente pela taxa de juros de mercado. 

Note que utilidades são valores subjetivos que devem ser trazidos a valor presente utilizando a taxa de impaciência (subjetiva) indivíduo. Recursos são valores que podem ser transacionados no mercado e são trazidos a valor presente utilizando a taxa de juros de mercado.

(iii) Essa condição indica que os preços sombra dos recursos, devidamente descontados, devem ser iguais. Em outras palavras, receber $100$ hoje ou $(1+r)100$ amanhã dá no mesmo; supondo, como estamos fazendo, que você possa emprestar ou pedir emprestado livremente à taxa de juros r.

(iv)-(v) indicam que a solução do problema deve respeitar as restrições. 

♣

### A equação de Euler

As equaões (i)-(ii)-(iii) implicam uma condição fundamental em qualquer problema de otimização intertemporal de consumo: a equação de Euler. 

<a id='voltar3'></a>
[Exercício resolvido 3](#Link3)

Mostre que as condições de ótimo (i)-(ii)-(iii) implicam:

$ \cfrac{c_2}{c_1}=\cfrac{1+r}{1+\rho} $

♦

Essa equação é muito importante e vamos interpretá-la de várias formas. Mas note que a equação de Euler relaciona duas variáveis endógenas $(C_1,C_2)$ e, portanto, não fornece a solução para nenhuma das duas. (Lembre que a solução para uma variável endógena só pode depender das exógenas.) Apesar de não ser a solução do problema, ela indica aspectos importantes da escolha ótima. Vamos às interpretações.

####  Perfil do consumo

$ \cfrac{c_2}{c_1}=\cfrac{1+r}{1+\rho} $

A expressão indica que quando a taxa de impaciência for igual à taxa de juros, o indivíduo escolherá um nível de consumo igual nos dois períodos. Se a impaciência for maior que os juros, o lado direito da equação será maior que um, e o indivíduo escolherá consumir mais no presente do que no futuro. Se a impaciência for menor que os juros, o lado direito da equação será menor que um, e o indivíduo escolherá consumir mais no futuro do que no presente.

#### Taxa marginal de substituição = preço relativo

$ \cfrac {\cfrac {\cfrac{1}{c_2}}{1+\rho}}{\cfrac{1}{c_1}} =\cfrac{1}{1+r} $

Do lado esquerdo temos a razão das utilidades marginais (descontadas), ou seja, a taxa marginal de substituição. Do lado direito, temos o preço relativo dos bens: para ter uma unidade hoje, devo abrir mão de $(1+r)$ unidades amanhã.

#### Custo marginal = benefício marginal

$ \cfrac{1}{c_1}=(1+r)\cfrac{1}{c_2}\cfrac{1}{1+\rho} $

Suponha que o indivíduo reduza seu consumo corrente em uma unidade, sua perda de satisfação será dada pela utilidade marginal do consumo hoje $\left( \cfrac{1}{c_1} \right)$. A unidade não consumida hoje será aplicada e renderá $(1+r)$ unidades de bens amanhã. Cada uma dessas unidades aumentará sua satisfação amanhã no valor da utilidade marginal do consumo amanhã $\left( \cfrac{1}{c_2} \right)$. Para compararmos a perda de utilidade hoje com o ganho amanhã, as duas variações precisam estar em unidades do mesmo período. Para isso, descontamos a variação na utilidade futura pela taxa de impaciência, ou seja, dividimos por $(1+\rho)$.

Dito de outra forma. Suponha que o indivíduo receba uma unidade adicional de recursos. Se ele a utilizar para consumir hoje obterá a utilidade marginal do consumo corrente. Se poupar, terá $(1+r)$ amanhã, que lhe darão uma utilidade marginal $\left( \cfrac{1}{c_1} \right)$, que deverá ser descontada por $(1+\rho)$.



### As escolhas ótimas

Agora vamos resolver o sistema (i)-(v) das condições de primeira ordem e obter as escolhas ótimas. A equação de Euler relaciona duas variáveis endógenas $(c_1,c_2)$ e, portanto, não é a solução para nenhuma delas. Para achar a solução para os consumos, é necessário combinar essa condição de primeira ordem e a restrição orçamentária intertemporal.

<a id='voltar4'></a>
[Exercício resolvido 4](#Link4) 

Resolva o sistema das condições de primeira ordem e mostre que as escolhas ótimas do consumidor são:

$ c_1 = \cfrac{1+\rho}{2+\rho} \left( y_1 + \cfrac{y_2}{1+r} \right) $

$ c_2 = \cfrac{1+r}{2+\rho} \left( y_1 + \cfrac{y_2}{1+r} \right) \Leftrightarrow
    \cfrac{c_2}{1+r} = \cfrac{1}{2+\rho}\left( y_1 + \cfrac{y_2}{1+r} \right) $

$ a_1 = \cfrac{y_1}{2+\rho} - \cfrac{1+\rho}{2+\rho}\cfrac{y_2}{1+r} $

♣

**Comentário**

(i) Cada escolha ótima só depende de variáveis exógenas (e parâmetros). A solução para uma variável endógena não pode depender de outras endógenas.

(ii) A expressão para o consumo ótimo do primeiro período depende de dois termos. O segundo termo entre parênteses  é o valor da riqueza total do agente ao longo da vida; é o valor presente do fluxo de renda, ou seja, o valor dos recursos em unidades do primeiro período. Note que o que importa é o valor da riqueza e não o perfil da renda. Nesse contexto, dois perfis de renda diferentes que tenham o mesmo valor presente são equivalentes. Há situações em que esse resultado pode não ser verdadeiro; por exemplo, quando há restrições de liquidez ou spread bancário.

O primeiro termo entre parênteses na expressão do consumo ótimo no primeiro período  é a propensão a consumir a partir da riqueza. Ele indica qual proporção da riqueza total do indivíduo ao longo da vida será alocada para consumo no primeiro período. No caso atual, essa propensão será um número no intervalo $(0,1)$. Quanto mais o indivíduo for impaciente – quanto maior  – maior será o valor da propensão a consumir. 

Se o indivíduo receber 10 adicionais no primeiro período, ele consumirá uma parte no primeiro período e poupará o resto. O mesmo ocorre se o indivíduo receber $10 no segundo período. No caso atual – em particular, sem restrição de liquidez –, o indivíduo sempre repartirá qualquer variação, positiva ou negativa, na disponibilidade de recursos em um período entre os consumos nos dois períodos. Com restrição de liquidez, isso pode não ser verdade.

O valor da propensão a consumir não depende da taxa de juros. Esse não é um resultado geral; ele depende de estarmos utilizando a função de utilidade logarítmica. No caso da função logarítmica, a taxa de juros só afeta o consumo do primeiro período através do valor da riqueza. Se a função de utilidade for logarítmica e o indivíduo não tiver renda no segundo período, o consumo ótimo do primeiro período não dependerá da taxa de juros.    

(iii) A interpretação da expressão para o consumo ótimo do segundo período é semelhante e isso fica mais evidente na segunda expressão que indica o valor presente do consumo futuro $\left( \cfrac{c_2}{1+r} \right)$. Note que a soma das propensões a consumir para os dois períodos é igual a um, como tem que ser.

(iv) Os dois consumos podem ser escritos como uma constante vezes a riqueza, ou seja, não dependem do perfil da renda, apenas do seu valor presente. A poupança não pode ser escrita dessa forma. Ela depende do perfil da renda e não apenas do seu valor presente. A poupança (ótima) depende do valor da renda e do consumo (ótimo). Caso a renda no período esteja acima do consumo desejado, o indivíduo poupa; caso contrário, ele se endivida. Resumindo, o consumo ótimo depende da riqueza e não do perfil da renda. A poupança permite que o indivíduo atinja esse perfil de consumo qualquer que seja o perfil da renda. (Lembre que estamos supondo que não haja restrições de crédito.)  

Um indivíduo que só tenha renda no primeiro período  será, necessariamente, poupador e um indivíduo que só tenha renda no segundo período  será, necessariamente, tomador de empréstimo. De maneira mais geral, a expressão da poupança indica que pessoas que atuam em áreas em que a maior parte da remuneração é obtida cedo na vida – exemplo: atletas de sucesso –, pouparão enquanto jovens. Da mesma forma, as pessoas tenderão a poupar para a velhice, quando sua capacidade de trabalho diminui acentuadamente. Por outro lado, pessoas envolvidas em um programa de estudos que aumente expressivamente suas perspectivas de renda futura tenderão a se endividar – caso consigam os empréstimos –, seja para financiar seu programa de estudos, seja para financiar seu consumo corrente.

## Política fiscal

Vamos introduzir gastos públicos e impostos na nossa economia simplificada.

Primeiro, vamos reescrever o problema do consumidor incluindo impostos. Supomos que o governo cobre impostos no valor de $t_1$  no primeiro período e $t_2$ no segundo período. Nesse caso, o problema do consumidor será:
 
$ \left\{ \begin{array}{l l}	
\underset{(c_1,c_2,a_1)}{max} U(c_1,c_2)  = ln(c_1) + \cfrac{ln(c_2)}{1+\rho}\\
     y_1 - t_1 = c_1 + a_1 \\
     y_2 - t_2 + (1+r)a_1 = c_2  \\
    \end{array}\right. $

Vamos supor, inicialmente, que os impostos sejam *lump-sum*, ou seja uma quantia fixa que o indivíduo deve pagar independente de qualquer escolha ou característica pessoais. Ou seja, o total de imposto pago não depende, por exemplo, nem da renda nem do consumo do indivíduo. Impostos *lump-sum* são raros na prática, mas são úteis para analisar uma situação em que queremos concentrar nos efeitos dos impostos sobre a renda disponível abstraindo de possíveis mudanças de preços relativos.

## Conclusão

(i) Ao tomarem suas decisões, as pessoas não olham apenas para o presente. Elas também levam em conta suas expectativas quanto o futuro. Na atual versão simplificada do modelo, não há incerteza e, portanto, as pessoas sempre acertam suas previsões. Mas isso é apenas uma simplificação.

(ii) Vimos como o consumidor reage a diversas mudanças no seu ambiente econômico. O procedimento envolve dois passos: calcular as escolhas ótimas, ou seja, escrever cada variável endógena apenas em função das exógenas; calcular a mudança, caso haja, na escolha ótima em função das mudanças no ambiente econômico.

(iii) O consumo ótimo do primeiro período depende do valor presente da renda em ambos os períodos e não apenas da renda no primeiro período. O mesmo vale para o consumo do segundo período. O indivíduo sempre utiliza seus recursos para consumir alguma coisa em cada um dos períodos. Portanto, uma variação na renda de um período provoca mudanças no consumo em ambos os períodos. E as variações na poupança servem para amortecer o impacto das variações da renda sobre o consumo.

(iv) O perfil do consumo ao longo do tempo depende da comparação entre a taxa de juros e a taxa de impaciência. Esse perfil será crescente, constante ou decrescente quando a taxa de juros for, respectivamente, maior, igual ou menor que a taxa de impaciência.

(v) A propensão a consumir não depende da taxa de juros. Esse não é um resultado geral, mas depende da função de utilidade ser logarítmica.

(vi) A inclusão de impostos *lump-sum* não altera a estrutura do modelo, exigindo apenas uma reinterpretação das variáveis. Embora outras modificações normalmente impliquem em maiores mudança, esse é um exemplo da flexibilidade do modelo básico.

**Exemplo 2**

A interpretação do multiplicador

Antes de resolver o problema acima, é útil vermos qual o significado econômico do multiplicador de Lagrange. Isso ajudará a interpretação econômica da solução do problema. Para isso, vejamos um problema de maximização de utilidade extremamente simplificado. O indivíduo maximiza a utilidade do consumo de dois bens $(x,y)$ dada uma quantidade de recursos $(z)$.

$ \left\{ \begin{array}{l l}	
\underset{(x,y)}{max} \,\,U(x,y)  = ln(x) + ln(y)\\
     x + y = z  \\
    \end{array}\right. $

O Lagrangeano do problema é:

$ \underset{(x,y,\lambda)}{max} \,\,L  = ln(x) + ln(y) + \lambda(z - x - y)   $

As condições de primeira ordem são:

$ \left\{ \begin{array}{l l}
\cfrac{\partial L}{\partial x} = 0 \Rightarrow\,\, \cfrac{1}{x} = \lambda \\
\cfrac{\partial L}{\partial y} = 0 \Rightarrow\,\, \cfrac{1}{y} = \lambda \\
\cfrac{\partial L}{\partial \lambda} = 0 \Rightarrow\,\,  z = x + y \\
\end{array}\right.$

É fácil ver que a solução é: $ x=y=\cfrac{1}{\lambda}=\cfrac{z}{2} $

A utilização ótima dos recursos $(z)$ é dada pela solução do problema de otimização: as funções de demanda por $x$ e $y$. Substituindo as soluções para $x$ e $y$ na função de utilidade direta $(U(x,y))$, obtemos a função de utilidade indireta $(V(z))$:

$ V(z) = ln\left(\cfrac{z}{2}\right)+ln\left(\cfrac{z}{2}\right) = 2(ln(z)-ln(2)) $

A função de utilidade indireta indica a satisfação máxima que pode ser obtida a partir de uma determniada quantidade exógena de recursos. Para calcularmos x o ganho máximo de satisfação com uma unidade de recursos adicional, basta derivar a função de utilidade indireta em relação aos recursos:

$ \cfrac{dV(z)}{dz} = \cfrac{2}{z} = \lambda $

Vemos que, no ponto de ótimo, o valor do multiplicador é igual à derivada da função de utilidade indireta em relação ao recurso exógeno. 
Portanto, o multiplicador indica o ganho de satisfação que teríamos se recebêssemos uma unidade a mais de recursos e a utilizássemos de maneira ótima. O multiplicador indica o valor, em termos de utilidade, de uma unidade a mais de recursos e, por isso, é conhecido como o "preço sombra" (*shadow price*) do recurso. E, embora tenhamos visto apenas um exemplo, essa interpretação é geral.

### Cálculo da riqueza em *N* períodos

A próxima célula calcula o valor da riqueza e os fatores de desconto dadas as rendas e as taxas de juros em um número arbitrário de períodos. O código segue os seguintes passos:

(i) Define uma função **WR** que tem como argumentos os vetores $y$ com as rendas e $r$ com as taxas de juros. A função começa em ```function``` e termina no último ```end```.

(ii) A função verifica se os vetores $y$ e $r$ têm o mesmo tamanho e envia uma mensagem de erro se não tiverem.

O *zero* na primeira posição do vetor $r$ é necessário para que ele tenha o mesmo tamanho que $y$. Podemos interpretá-lo como a taxa de juros entre *hoje* e *hoje*.

O comando ```length(x)``` calcula o tamanho de *x*. O comando ```x != y``` significa *x diferente de y*. Veja [aqui](https://en.wikibooks.org/wiki/Introducing_Julia/Arrays_and_tuples).

O comando ```println("texto")``` escreve uma mensagem na tela. Veja [aqui](https://en.wikibooks.org/wiki/Introducing_Julia/print).

(iii) Calcula um valor provisório para os fatores de desconto $R$. Isso é necessário para que o *loop* que virá depois tenha um ponto de partida.

Note o *ponto* antes das operações matemáticas.

(iv) Define um [loop](https://docs.julialang.org/en/v1/manual/control-flow/index.html) para calcular os valores definitivos de $R$. O *loop* começa em ```for``` e termina no primeiro ```end```. O *loop for* repete um conjunto de comandos um número pré-definido de vezes.

(v) Define a saída da função **WR** que será composta do escalar $W$ e do vetor $R$.

(vi) Fora da função **WR** são definidos os valores para $y$ e $r$. Esses comandos poderiam vir antes da definição da função, mas têm que vir antes da função ser chamada.

(vii) Chama a função **WR** com os valores estabelecidos para os parâmetros.


````julia
function WR(y,r)                 # início da função WR

    # Verificação das dimensões dos vetores
    
    Ny = length(y)
    Nr = length(r)
    
    if Ny != Nr                                 # verifica se os vetores têm o mesmo tamanho
        println("dimensões incompatíveis")      # envia mensagem de erro, se for o caso
    end
    
    ## Cálculo dos fatores de desconto
    
    R = 1 ./(1 .+ r)            # valores provisórios para R
 
    for k = 2:Ny                
        R[k] = R[k]*R[k-1]      # valores definitivos para R
    end
        
    # Cálculo da riqueza
    
    W = y*R'
    
    # Definição da saída da função
    
    WR = W, R
    
    end                         # final da função WR

y = [100 220 121]
r = [0 0.1 0.1]

WR(y,r) 
````
