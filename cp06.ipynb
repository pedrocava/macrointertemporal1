## Introdução

Até agora fizemos de conta que o futuro fosse previsível. Essa é uma simplificação útil para analisar vários problemas, mas impede a discussão de outros. Sem incerteza, o modelo implica uma relação exata entre as variáveis exógenas e as endógenas, o que não é compatível com as características dos ciclos econômicos vistas anteriormente.

Além disso, vários eventos incertos ocorrem de forma recorrente e apresentam alguma regularidade que pode ser percebida pelas pessoas. Por exemplo, os retornos dos investimentos variam ao longo do tempo e é possível analisá-los em termos de retorno esperado e risco. Nossa economia simplificada ainda não é capaz de lidar com esse componente da tomada de decisões. Neste capítulo, veremos como incluir a incerteza quanto ao futuro em nossa economia simplificada.


## O modelo com incerteza

Para manter uma estrutura simples, vamos supor que a única fonte de incerteza seja o valor da renda futura. Como isso altera o problema original, repetido abaixo? 

$ \left\{ \begin{array}{l l}	
\underset{(C_1,C_2,A_1)}{max} U(C_1,C_2)  = ln(C_1) + \cfrac{ln(C_2)}{1+\rho}\\
     Y_1 = C_1 + A_1 \\
     Y_2 + (1+r)A_1 = C_2  \\
    \end{array}\right. $
    
Vejamos as modificações necessárias no nosso modelo

**Modificação 1**

O primeiro passo é especificar as características das variáveis aleatórias exógenas. No nosso exemplo, apenas a renda é incerta. E, no caso mais simples possível, pode  assumir  apenas  dois  valores: $Y^G$ e $Y^P$,  com  probabilidades $p$ e $1-p$. (Dizemos que há apenas dois estados da natureza possíveis no segundo período.) Isso caracteriza completamente a distribuição de probabilidade de $Y$: listamos todos os seus possíveis valores e suas respectivas probabilidades de ocorrência. 

♠


**Modificação 2**

O segundo passo é especificar os conjuntos de informação do indivíduo: o que ele sabe a cada momento. Se você comprar um bilhete em uma loteria honesta, só saberá o resultado após o sorteio. Antes do sorteio, o número sorteado não faz parte do seu conjunto de informações; depois do sorteio, ele faz parte. É claro que suas decisões ótimas quanto, por exemplo, ao seu nível de consumo dependem do seu conjunto de informação. 

Vamos representar os conjuntos de informação do indivíduo da seguinte forma:

$ℑ_1$: tudo aquilo que o indivíduo sabe no primeiro período. 

$ℑ_2$: tudo aquilo que o indivíduo sabe no segundo período, possivelmente, incluindo aquilo que ele já sabia no primeiro período. 

(O conjunto de informação é comumente representado pela letra maiúscula $I$. Para distingui-lo do investimento, usa-se um $I$ modificado: $ℑ$.) 

Quais os elementos de $ℑ_1$ e $ℑ_2$?

$ℑ_1: (\rho,\, r,\, Y_1,) $

$ℑ_2: (\rho,\, r,\, Y_1,\, Y_2) = (ℑ_1,\, Y_2) $ 

Portanto, a única informação que o indivíduo recebe no segundo período é o valor da tecnologia deste período. O indivíduo toma suas decisões em cada período com base no conjunto de informação daquele período. Por exemplo, o indivíduo escolhe $(C_1)$ no primeiro período com o conjunto de informação $ℑ_1$; e escolhe $C_2$ no segundo período com o conjunto de informação $ℑ_2$. 

Vale a pena insistir nesse ponto. $(C_1)$ é uma variável endógena que o indivíduo escolhe no primeiro período e, portanto, não há incerteza quanto ao seu valor; o indivíduo sabe o que está fazendo, o que ele não sabe, no primeiro período, é o que escolherá no segundo período, $C_2$. É claro que quando o segundo período chegar, $C_2$ deixará de ser uma variável aleatória.

♣

**Modificação 3**

Precisamos reconsiderar o objetivo do indivíduo. Quando há incerteza quanto ao valor da renda futura, o indivíduo não é mais capaz de maximizar sua utilidade ao longo da vida inteira porque, quando decide quanto consumir hoje, ainda não sabe o valor da renda futura. Como vimos anteriormente, se ele sabe que estará rico (pobre) no segundo período, desejará consumir muito (pouco) já no primeiro período. 

Sem saber o valor da renda futura, um critério razoável é tentar maximizar sua expectativa de utilidade ao longo da vida – o valor médio da utilidade levando em conta a média dos valores possíveis para a tecnologia. Supondo que o indivíduo maximize sua utilidade esperada, essa parte do problema passa a ser:

$ 	
\underset{(C_1,C_2,A_1)}{max} E_1[U(C_1,C_2)]  = E_1 \left[ ln(C_1) + \cfrac{ln(C_2)}{1+\rho} \right]\\
$

Isso pode ser reescrito da seguinte forma:

$E_1[U(C_1,C_2)]  = E_1 \left[ ln(C_1) + \cfrac{ln(C_2)}{1+\rho} \right] =
    E_1 \left[ ln(C_1) \right] + E_1 \left[ \cfrac{ln(C_2)}{1+\rho} \right] =
    ln(C_1) + E_1 \left[ \cfrac{ln(C_2)}{1+\rho} \right] = \\
    \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
    = ln(C_1) + \cfrac{ E_1 \left[ln(C_2) \right]}{1+\rho} = 
    ln(C_1) + \cfrac{ p.ln(C_2^G) + (1-p).ln(C_2^P) }{1+\rho}  
$

A primeira igualdade vale porque o operador de expectativas é linear. (A expectativa de uma variável aleatória é a média ponderada dos valores possíveis da variável, onde os pesos são as probabilidades de ocorrência de cada valor. Por isso, a expectativa da soma é igual à soma das expectativas.) A segunda igualdade vale porque a renda do primeiro período é conhecida no momento da tomada de decisão e sai do operador de expectativa. A terceira igualdade vale porque a impaciência é constante e também sai do operador de expectativa. E a quarta igualdade vale porque, no segundo período, $u(C_2)=ln(C_2^G)$, com probabilidade $p$, e $u(C_2)=ln(C_2^P)$, com probabilidade $(1-p)$.

(Note que $E_1[ln(C_2)) \neq ln(E_1[C_2])$, pois a função logarítmica não é linear. Você pode verificar a desigualdade com uma calculadora.)

♦



**Modificação 4**

Vejamos as restrições orçamentárias. É importante notar que a restrição orçamentária do segundo período deve ser respeitada quer a renda do segundo período seja grande ou pequena. Dito de outra forma, a restrição orçamentária precisa ser respeitada em cada estado da natureza (*ex-post*) e não apenas na média (*ex-ante*). É claro que sendo sempre respeitada, ela será respeitada na média, mas a volta não é verdadeira.  Portanto, precisamos de uma restrição para cada período e para cada estado da natureza. 

No primeiro período não há incerteza e temos apenas uma restrição. No segundo período há dois estados da natureza possíveis e, portanto, duas restrições, uma para cada valor da tecnologia.

$ \left\{ \begin{array}{l l}	
     Y_1 = C_1 + A_1 \\
     Y_2^G + (1+r)A_1 = C_2^G  \\
     Y_2^P + (1+r)A_1 = C_2^P  \\
    \end{array}\right. $

Como o único ativo disponível tem o mesmo valor em ambos os estados da natureza, estamos na situação que é chamada de [mercados incompletos](https://en.wikipedia.org/wiki/Incomplete_markets). O indivíduo pode transferir recursos ao longo do tempo, mas não tem como transferir recursos entre estados da natureza. Se, no segundo período, o valor da renda for grande, ele terá muitos recursos e poderá consumir muito. Se for pequeno, deverá consumir menos. 

Nesse exemplo, precisaríamos de dois ativos independentes, um para cada estado da natureza, para termos [mercados completos](https://en.wikipedia.org/wiki/Complete_market). Um seguro contra acidentes, por exemplo, permite transferir recursos entre estado da natureza.

♥

Com as modificações acima, nosso problema passa a ser:

$ \left\{ \begin{array}{l l}	
\underset{(C_1,C_2^G,C_2^P,A_1)}{max} E_1[U(C_1,C_2)]  = ln(C_1) + \cfrac{ p.ln(C_2^G) + (1-p).ln(C_2^P) }{1+\rho} \\
     Y_2^G + (1+r)A_1 = C_2^G  \\
     Y_2^P + (1+r)A_1 = C_2^P  \\
    \end{array}\right. $

Note a existência de duas restrições orçamentárias intertemporais, uma para cada estado da natureza: o indivíduo tem que viver dentro dos recursos de que efetivamente dispõe. No problema sem incerteza, havia apenas uma restrição orçamentária intertemporal.

**Comentário**

(i) As condições de primeira ordem são versões modificadas das condições equivalentes sem incerteza. Fazendo $Y^G=Y^P$ ou $p=\{0\,;1\}$  obtemos os resultados sem incerteza. 

(ii) Poderíamos considerar uma função de utilidade genérica, incerteza quanto à taxa de juros e escrever e um número arbitrário de períodos. Nesse caso mais, a equação de Euler será:

$ u'(C_t) = \cfrac{1}{1+\rho}\, E_t \left[ (1+r) \, u'(C_{t+1}) \right] $

(iii) Para respeitar as restrições orçamentárias, o indivíduo continua tendo apenas um grau de liberdade. Escolhido o consumo do primeiro período, ele consumirá o que for possível no segundo período, dependendo do valor da renda. 

(iv) Embora o modelo seja muito simples, não há como obter uma solução analítica.

## Equilíbrio

O cálculo do equilíbrio segue a mesma linha do caso da economia sem incerteza.

As condições de equilíbrio na pequena economia aberta correspondem às condições de ótimo para o agente representativo, vistas na última seção: 

$ \left\{ \begin{array}{l l}	
    \cfrac{1}{C_1} = \cfrac{1+r}{1+\rho} \left[ \cfrac{p}{C_2^G} + \cfrac{1-p}{C_2^P} \right] 
       = \cfrac{1+r}{1+\rho} E_1 \left[ \cfrac{1}{C_2} \right]\\
    Y_1 = A_1 + C_1 \\
     Y_2^G + (1+r)A_1 = C_2^G  \\
     Y_2^P + (1+r)A_1 = C_2^P  \\
    \end{array}\right. $

Mostre que na autarquia a taxa de juros de equilíbrio é dada implicitamente por:

$ 
\cfrac{1}{Y_1} = \cfrac{1+r}{1+\rho} \left[ \cfrac{p}{Y_2^G} + \cfrac{1-p}{Y_2^P} \right] =
    \cfrac{1+r}{1+\rho} E_1 \left[ \cfrac{1}{Y_2} \right] \\
$

♦

**Comentário**

(i) A equação que determina a taxa de juros de equilíbrio na autarquia com incerteza é uma extensão imediata da equação equivalente no caso sem incerteza. A diferença é que a utilidade marginal da renda do segundo período é substituída pela utilidade marginal da renda *esperada* do segundo período.

(ii) $ E_1 \left[ \cfrac{1}{Y_2} \right] \neq  \cfrac{1}{E_1 \left[ Y_2 \right] }$

♠

## Exemplo

Primeiro, vamos resolver o problema do indivíduo supondo $Y^G=Y^P=100$, ou seja, sem incerteza. Os demais parâmetros são: $\rho=r=0.05\,,\,p=0.5$. (O valor de p é irrelevante neste caso.) A solução é:

$ [C1 \;\; C_2^G \;\; C_2^P \;\; A_1] = [100.0 \;\; 100.0 \;\; 100.0 \;\; 0.0] $

O consumo é igual nos dois períodos, pois a taxa de juros é igual à de impaciência. 

Suponha agora que $Y^G=110$ e $Y^P=90$. O valor esperado da renda não mudou, mas sua dispersão aumentou. A solução agora é:

$ [C1 \;\; C_2^G \;\; C_2^P \;\; A_1] = [99.5147 \;\; 110.51 \;\; 90.5096 \;\; 0.485332] $

O aumento na incerteza leva o indivíduo a poupar no primeiro período para evitar um consumo muito pequeno no segundo período caso a renda seja pequena. Um indivíduo avesso ao risco reage ao aumento na dispersão da renda futura de maneira semelhante ao que um indivíduo neutro ao risco faria frente a uma queda no valor esperado da renda futura.

Note que se o indivíduo não poupasse, teríamos:

$ u'(C_1) = \cfrac{1}{Y_1} = 0.01 \neq 0.0101 = 0.5. \left( 1/110 + 1/90 \right) = 
    \left( \cfrac{p}{Y_2^G} + \cfrac{1-p}{Y_2^P} \right) = E_1 \left[ u'(C_2) \right] $

Suponha agora que a economia seja uma autarquia. O aumento no risco da renda futura levará uma queda na taxa de juros. O mecanismo é claro: os indivíduos desejam poupar mas, agregadamente, isso não é possível na nossa autarquia. O resultado é a queda nos juros até que os indivíduos decidam consumir aquilo que está disponível em cada período e em cada estado da natureza. O juro de equilíbrio será: 0.0395.

## Expectativas racionais

O modelo utiliza a [hipótese de expectativas racionais](https://www.econlib.org/library/Enc/RationalExpectations.html), comumente abreviada como *REH*. 

Informalmente, *REH* assume que os erros de previsão sobre as variáveis endógenas não estão correlacionados com o conjunto de informação do indivíduo. Se você, sistematicamente, sobrestima o resultado do seu time quando ele joga fora de casa, você não está fazendo um uso eficiente da informação sobre o mando de campo. Se você estiver apostando e não quiser perder denheiro, vai rever para baixo suas previsões.

*REH* implica que a distribuição de probabilidade que os agentes atribuem à endógenas é a mesma produzida pela solução do modelo. Essa é uma aproximação razoável. Digamos que você: (i) suponha que o câmbio irá desvalorizar; (ii) utilize seu modelo de comércio exterior e conclua que haverá um enorme superávit comercial; (iii) que o superávit irá provocar uma valorização do câmbio. Algo está, obviamente, errado. Uma expectativa de desvalorização do câmbio não pode implicar uma expectatia de valorização do câmbio. 

Em um modelo com expectativa racionais, as expectativas dos agentes são endógenas. E são calculadas utilizando as mesmas distribuições de probabilidade que a solução do modelo atribui àquelas variáveis. 

*REH* não deve ser confundida com previsão perfeita. Previsão perfeita implica ausência de erros de previsão. *REH* é uma hipótese mais fraca que implica que a média dos erros de previsão é zero. Se não houver incerteza, os dois conceitos são equivalentes.

Como todo elemento de modelagem, expectativas racionais são uma simplificação. Mas essa é uma simplificação útil ou enganosa? A hipótese é empiricamente plausível?


Em situações moderadamente estáveis, parece razoável supor que as pessoas percebam as regularidades empíricas. As pessoas têm interesse econômico em fazer boas previsões e há milhões delas tentando e se comunicando umas com as outras. Um exemplo extremo é a regularidade de que nos meses de verão a temperatura média é superior à dos meses de inverno.

E em situações mais complexas? Quanto mais a estrutura da economia estiver mudando - mais frequente e acentuadamente - mais difíceis serão as previsões. É provável que, logo após mudanças drásticas na economia, haja um período de aprendizado durante o qual expectativas racionais não sejam uma boa aproximação. Mas, à medida em que as pessoas façam um uso eficiente da informação disponível, aprendam sobre a nova situação e corrijam seus erros, gradualmente a aproximação da *REH* se torna mais convincente.

## Conclusão

Neste capítulo vimos como expandir nossos modelos determinísticos para incluir incerteza. A maneira como representamos a existência de dois estados da natureza é semelhante à que usamos para representar a existência de dois períodos. Só existe um bem físico na economia. Mas um bem físico hoje representa um bem econômico diferente do mesmo bem físico amanhã. Da mesma forma, um bem físico em um estado da natureza favorável representa um bem econômico diferente do mesmo bem físico em um estado da natureza desfavorável. Essa semelhança na maneira de representar os problemas é cômoda, pois permite utilizar a mesma estrutura analítica em ambos os casos.

     Y_2^P + (1+r)A_1 = C_2^P  \\
    \end{array}\right. 
\\ $

$ \left\{ \begin{array}{l l}	
    \cfrac{1}{C_1} = \cfrac{1+r}{1+\rho} \, E_1 \left[ \cfrac{1}{C_2} \right] 
        = \cfrac{1+r}{1+\rho} \, \left( \cfrac{p}{C_2^G} + \cfrac{1-p}{C_2^P} \right) \\
        Y_1 = C_1 + A_1 \\
     Y_2^G + (1+r)A_1 = C_2^G  \\
     Y_2^P + (1+r)A_1 = C_2^P  \\
    \end{array}\right. 
\\ $

$ \left\{ \begin{array}{l l}	
    \cfrac{1}{C_1} = \cfrac{1+r}{1+\rho} \, E_1 \left[ \cfrac{1}{C_2} \right] \\
    C_1 = Y_1 \\
     C_2^G = Y_2^G \\
     C_2^P = Y_2^P \\
    \end{array}\right. $

**Exercício resolvido 1**

Problema do indivíduo

$ \left\{ \begin{array}{l l}	
\underset{(C_1,C_2^G,C_2^P,A_1)}{max} E_1[U(C_1,C_2)]  = ln(C_1) + \cfrac{ p.ln(C_2^G) + (1-p).ln(C_2^P) }{1+\rho} \\
     Y_2^G + (1+r)A_1 = C_2^G  \\
     Y_2^P + (1+r)A_1 = C_2^P  \\
    \end{array}\right. 
\\ $

Lagrangeano

 $   \underset{X}{max} L = ln(C_1) + \cfrac{ p.ln(C_2^G) +  
     (1-p).ln(C_2^P) }{1+\rho} + \lambda_1(Y_1 - C_1 - A_1) + \\
     \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
     + \, \lambda_2^G[Y_2^G + (1+r)A_1 - C_2^G]  + \lambda_2^P[Y_2^P + (1+r)A_1 - C_2^P] \\
     \\ \; \\
     X = (C_1,C_2^G,C_2^P,A_1,\lambda_1,\lambda_2^G,\lambda_2^P) \\
\\ $

        
Condições de primeira ordem

$ \left\{ \begin{array}{l l}
(i)\,\, \cfrac{\partial L}{\partial C_1} = \cfrac{1}{C_1} - \lambda_1 = 0 \,\, \Rightarrow\,\, \cfrac{1}{C_1} = \lambda_1  \\
(ii)\,\, \cfrac{\partial L}{\partial C_2^G} = \cfrac{p}{1+\rho} \cfrac{1}{C_2^G} - \lambda_2^G = 0 \,\,  \Rightarrow\,\,            \cfrac{1}{1+\rho} \cfrac{1}{C_2^G} = \lambda_2^G\\
(iii)\,\, \cfrac{\partial L}{\partial C_2^P} = \cfrac{1-p}{1+\rho} \cfrac{1}{C_2^P} - \lambda_2^P = 0 \,\,  \Rightarrow\,\,            \cfrac{1}{1+\rho} \cfrac{1}{C_2^P} = \lambda_2^P\\
(iv)\,\, \cfrac{\partial L}{\partial A_1} = - \lambda_1 + (1+r)\,(\lambda_2^G + \lambda_2^P) = 0 \,\,  
    \Rightarrow\,\, \lambda_1 = (1+r)\,(\lambda_2^G + \lambda_2^P)  \\
(v)\,\, \cfrac{\partial L}{\partial \lambda_1} = Y_1 - C_1 - A_1 = 0 \,\,  \Rightarrow\,\, Y_1 = C_1 + A_1  \\
(vi)\,\, \cfrac{\partial L}{\partial \lambda_2^G} = Y_2^G + (1+r)A_1 - C_2^G = 0 \,\, 
    \Rightarrow\,\, Y_2^G + (1+r)A_1 = C_2^G  \\
(vii)\,\, \cfrac{\partial L}{\partial \lambda_2^P} = Y_2^P + (1+r)A_1 - C_2^P = 0 \,\, 
    \Rightarrow\,\, Y_2^P + (1+r)A_1 = C_2^P  \\
\end{array}\right. 
\\ $

Combinando (i)-(ii)-(iii)-(iv), obtemos:

$ \left\{ \begin{array}{l l}
(i)\,\, \cfrac{1}{C_1} = \color{blue}{\lambda_1}  \\
(ii)\,\, \cfrac{1}{1+\rho} \cfrac{1}{C_2^G} = \color{red}{\lambda_2^G} \\
(iii)\,\, \cfrac{1}{1+\rho} \cfrac{1}{C_2^P} = \color{magenta}{\lambda_2^P}\\
(iv)\,\, \color{blue}{\lambda_1} = (1+r)\,(\color{red}{\lambda_2^G} + \color{magenta}{\lambda_2^P})  \\
\end{array}\right. 
\;\;\;\;\;\; \Rightarrow \;\;\;\;\;\;
\cfrac{1}{C_1} = \cfrac{1+r}{1+\rho} \, \left( \cfrac{p}{C_2^G} + \cfrac{1-p}{C_2^P} \right)
$

## Anexo

Neste anexo veremos a solução de um modelo muito simples de expectativas racionais. O objetivo é apenas ilustrar como a *REH* garante a consistência entre as expectativas dos indivíduos e a solução do modelo. O exemplo não é representativo do grau de dificuldade envolvido na solução de modelos mais complexos.

### Economia

Em muitas situações, os preços de ativos dependem das expectativas quanto aos seus preços futuros. Por exemplo, o nível do câmbio hoje depende da expectativa de câmbio amanhã. Suponha que a estrutura da economia seja dada pela equação abaixo:

$ x_t = \alpha + \beta x_t^{t+1} + \epsilon_t $

Onde: $x_t$: endógena; $x_t^{t+1}$: expectativa de $x_{t+1}$ com base nas informações em *t*; $\epsilon_t \, \text{~} \, iid\,(0,\sigma_\epsilon^2)$.




### Expectativas exógenas

Com expectativas exógenas, $x_t^{t+1}$ é um dado do problema. Queremos verificar se ele é compatível com a expectativa de $x_{t+1}$ dada pela solução do modelo $(E_t[x_{t+1}])$.

Se $ x_t = \alpha + \beta x_t^{t+1} + \epsilon_t $, então:

$ x_{t+1} = \alpha + \beta x_{t+1}^{t+2} + \epsilon_{t+1} $ 

$ E_t[x_{t+1}] = E_t \left[ \alpha + \beta x_{t+1}^{t+2} + \epsilon_{t+1} \right] $

$ E_t[x_{t+1}] = \alpha + \beta x_{t+1}^{t+2}  \neq x_{t+1}^e $

Ou seja, nossa expectativa calculada utilizando nosso próprio modelo é imcompatível com  a nossa própria expectativa!

### Expectativas racionais

Com expectativas racionais, as expectativas sobre as variáveis endógenas são, elas próprias, endógenas e calculadas utilizando o modelo: $x_t^{t+1}=E_t[x_{t+1}]$. Ou seja a estrutura da economia é:

$ x_t = \alpha + \beta E_t[x_{t+1}] + \epsilon_t $

Onde $x_t$ e $E_t[x_{t+1}]$ são variáveis endógenas cujas soluções precisam ser calculadas. 

Pode parecer estranho querer resolver uma equação com duas endógenas. Mas note que existe uma relação implícita entre $x_t$ e $x_{t+1}$, pois nossa equação vale para todos os períodos. Adiantando um período, temos uma equação com $x_{t+1}$ do lado esquerdo *etc*.

Vamos resolver o problema pelo método dos coeficientes indeterminados que envolve duas etapas: (i) um palpite sobre qual será a solução; (ii) encontrar as condições para que o palpite esteja correto.

O palpite pode ser qualquer coisa, mas um palpite razoável facilita a vida. Lembre que: (i) a solução de uma equação só pode depender das exógenas, no caso, a constante e $ \epsilon_t $; (ii) uma equação linear terá uma solução linear. Dessa forma, um palpite razoável para a solução é: 

$ x_t = \phi_0 + \phi_1 \epsilon_t $

Onde $ ( \phi_0 , \phi_1 ) $ são coeficientes cujos valores ainda precisam ser determinados (daí o nome do método).

Dado o palpite, precisamos explorar suas implicações.


$  
\left\{ \begin{array}{l l}
x_t = \phi_0 + \phi_1 \epsilon_t \\
\\
x_{t+1} = \phi_0 + \phi_1 \epsilon_{t+1} \\
\\
E_t[x_{t+1}] = E_t \left[\phi_0 + \phi_1 \epsilon_{t+1} \right] = \phi_0 \\
\end{array} \\
\\ \right. $

Para que o palpite esteja correto é necessário que, ao substituí-lo na equação original, obtenhamos uma identidade. Substitua $ \left( x_t,E_t[x_{t+1}] \right) $ na equação original:

$ 
\left\{ \begin{array}{l l}
\color{red}{x_t} = \alpha + \beta \color{blue}{E_t[x_{t+1}]} + \epsilon_t \\
\\
\color{red}{x_t} = \phi_0 + \phi_1 \epsilon_t \\
\\
\color{blue}{E_t[x_{t+1}]} = \phi_0 \\
\end{array}  
\;\;\; \Rightarrow \;\;\; 
\phi_0 + \phi_1 \epsilon_t = \alpha + \beta \phi_0 + \epsilon_t \;\; (1)
\\ \right. $

A equação (1) deve ser verdadeira para quaisquer valores de $( \alpha, \beta, \epsilon_t )$. Para garantir isso, precisamos igualar os termos semelhantes dos dois lados da equação:

$ \\
\left\{ \begin{array}{l l} 
\text{cste}: \;\; \phi_0  + \phi_1 \epsilon_t = \alpha + \beta \phi_0 + \epsilon_t 
    \;\; \Rightarrow \;\;  \phi_0 = \cfrac{\alpha}{1-\beta} \\
\epsilon_t: \;\; \phi_1 \epsilon_t = \epsilon_t \;\; \Rightarrow \;\; \phi_1 =1 \\
\end{array} \\
\\ \right. $

Os valores de $(\phi_0,\phi_1)$ foram determinados e podemos substituí-los no palpite original.

$ \\
\left\{ \begin{array}{l l}
x_t = \color{blue}{\phi_0} + \color{red}{\phi_1} \epsilon_t \\
\\
\color{blue}{\phi_0} = \cfrac{\alpha}{1-\beta} \\
\\
\color{red}{\phi_1} = 1 \\
\end{array}  
\;\;\; \Rightarrow \;\;\; 
x_t = \cfrac{\alpha}{1-\beta} +  \epsilon_t 
\\  \right. $

Finalmente, verificamos que as expectativas dos indivíduos são consistentes com a solução do modelo. 

$ \\
\left\{ \begin{array}{l l}
x_t = \cfrac{\alpha}{1-\beta} +  \epsilon_t \\
\\
x_{t+1} = \cfrac{\alpha}{1-\beta} +  \epsilon_{t+1} \\
\\
E_t[x_{t+1}] = E_t \left[ \cfrac{\alpha}{1-\beta} +  \epsilon_{t+1} \right] 
    = \cfrac{\alpha}{1-\beta} = \phi_0 
\end{array} \\
\\ \right. $

### Problema econômico

No nosso problema específico, as equações relevantes são as que determinam o equilíbrio.

O equilíbrio na pequena economia aberta determina os consumos e a poupança:

$ \left\{ \begin{array}{l l}	
    \cfrac{1}{C_1} = \cfrac{1+r}{1+\rho} \, E_1 \left[ \cfrac{1}{C_2} \right] 
        = \cfrac{1+r}{1+\rho} \, \left( \cfrac{p}{C_2^G} + \cfrac{1-p}{C_2^P} \right) \\
        Y_1 = C_1 + A_1 \\
     Y_2^G + (1+r)A_1 = C_2^G  \\
     Y_2^P + (1+r)A_1 = C_2^P  \\
    \end{array}\right. 
\\ $

O equilíbrio na autarquia determina a taxa de juros:

$\cfrac{1}{Y_1} = \cfrac{1+r}{1+\rho} \left[ \cfrac{p}{Y_2^G} + \cfrac{1-p}{Y_2^P} \right]$

A próxima célula apresenta o código para calcular os equilíbrios. 

No código, as variáveis endógenas serão armazenadas no vetor $x$, cujo número de posições dependerá do problema.

Na pequena economia aberta: $x[1] = C1\,,\, x[2]=C_2^G\,,\, x[3]=C_2^P\,,\, x[4]=A_1$. Na autarquia: $x[1]=r$.

### Código

```julia
# Esta célula  carrega o pacote NLsolve e importa o solver nlsolve. 
# Ela só precisa ser executada uma vez.

Pkg.add("NLsolve")
import NLsolve: nlsolve
```

```julia

# Exógenas e parâmetros

Y1 = 100; Y2G = 100;  Y2P = 100; ρ = 0.05; p = 0.5;

# Pequena economia aberta
r  = 0.05; 

# Este bloco define o sistema de equações a ser resolvido.
# As equações correspondem ao equilíbrio na pequena economia aberta.
# x[1] = C1, x[2]=C2G, x[3]=C2P, x[4]=A1.      

f(x) = [ 1/x[1] - ((1+r)/(1+ρ))*(p/x[2]+(1-p)/x[3])
        Y1 - x[1] - x[4] 
        Y2G + (1+r)x[4] - x[2]
        Y2P + (1+r)x[4] - x[3] ]# returns an array


# Este bloco utiliza o solver nlsolve
# f é o sistema definido acima. 
# [100; 100; 100; 0.0] é o vetor com os palpites iniciais para as endógenas.

results = nlsolve(f, [ 100; 100; 100; 0.0])
x = results.zero

# Resultados

C1 = x[1];
C2G = x[2];
C2P = x[3];
A1 = x[4];

[C1 C2G C2P A1]
```

```julia
# Autarquia

Y1 = 100; Y2G = 110;  Y2P = 90; ρ = 0.05; p = 0.5;


# Este bloco define o sistema de equações a ser resolvido; no caso apenas uma.
# A equação corresponde ao equilíbrio na autarquia.
# x[1] = r.

f(x) = [ 1/Y1 - ((1+x[1])/(1+ρ))*(p/Y2G+(1-p)/Y2P) ]# returns an array

# Este bloco utiliza o solver nlsolve
# f é o sistema definido acima.

results = nlsolve(f, [ρ])
x = results.zero

# Resultados

r = x[1]

```
