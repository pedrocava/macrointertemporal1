## Introdução

Alguns dos objetivos da análise macroeconômica são: (i) entender como as principais variáveis econômicas se comportam ao longo do ciclo de negócios; (ii) como as economias reagem às mudanças no ambiente econômico; (iii) qual o efeito das políticas econômicas. Neste capítulo vamos começar com o primeiro item e analisar as características empíricas de algumas séries relevantes.

Começaremos a análise olhando para poucas séries: produção e consumo para Argentina e Brasil. Vamos discutir características como persistência e  volatilidade de cada série. E, também, a correlação entre as séries de produção e consumo para cada país. Esta é uma etapa importante: entender as características do problema que vamos analisar.

Este capítulo segue Uribe & Schmitt-Grohé [Open Economy Macroeconomics](https://press.princeton.edu/titles/11032.html). A construção dos dados está descrita na seção 4 deste capítulo; por exemplo, como foi construído o componente cíclico das séries.

### Correlação contemporânea

Em quase todos os países analisados por Uribe & Schmitt-Grohé, o consumo é pró-cíclico, ou seja, apresenta uma correlação positiva com a produção corrente: $cor(y_t,c_t)>0$. Nos nossos dados, a correlação entre os componentes cíclicos do *PIB* e do consumo é de 0,51 para o México e de 0,93 para o Brasil.


### Persistência

Informalmente, a persistência de uma série indica o quanto seu valor presente tende a ser parecido com os vizinhos. Uma medida de persistência é a correlação de primeira ordem: $cor(y_t,y_{t-1})$. Em séries macroeconômicas, esses valores tendem a ser elevados, , próximos de um, fazendo com que choques exógenos tenham efeitos duradouros. A correlação de primeira ordem para o componente cíclico do *PIB* é de 0,94 para o Brasil e de 0,85 para o México. A persistência do consumo é quase igual à da renda no Brasil (0,93) enquanto que no México é bem mais baixa (0,63).

### Tendência

Definimos o componente cíclico das séries como a diferença entre o valor observado e a estimativa de uma tendência quadrática. A primeira coisa a notar é que o componente cíclico foi construído, ele nunca é observado. Por exemplo, se tivéssemos utilizado uma tendência linear, teríamos outro componente cíclico. 

Note que esse não é um problema de falta de dados. Algumas atividades econômicas podem ser difíceis de medir. Por exemplo, o nível de atividade da economia informal. Mas mesmo que observássemos todas as transações econômicas, tudo o que teríamos seria uma medida do nível de atividade. Jamais poderíamos medir a *tendência* da série. Ela precisa ser construída. E há várias maneiras de fazê-lo. 

Discutir as alternativas para estimação da tendência de uma série envolveria conceitos de análise de séries temporais e nos forçaria a um longo desvio na discussão. Por isso, vamos apenas ilustrar que a definição da tendência pode ter um grande impacto na definição do componente cíclico. 

Vimos acima como estimar uma tendência quadrática. Para estimar uma tendência linear, procedemos da mesma forma, obviamente, eliminando o termo quadrático da regressão. A figura abaixo, compara os componentes cíclicos do PIB brasileiro construídos com a tendência quadrática vista acima e com uma tendência linear.



Vemos que a diferença entre os dois componentes cíclicos é bem maior no caso do Brasil que no da Argentina. Isso não surprende, visto que a tendência quadrática estimada para a Argentinha é muito mais próxima de linear que a brasileira.

### Duráveis e não-duráveis

Quando um modelo macroeconômico fala de *consumo* quase sempre está se referindo ao consumo de não-duráveis. Do ponto de vista econômico, os gastos com o consumo de duráveis representam a aquisição de um estoque que gera serviços ao longo do tempo. Portanto, trata-se de uma atividade de investimento. 

Infelizmente, em muitos casos, as séries econômicas disponíveis não desagregam o consumo entre duráveis e não-duráveis. E isso dificulta o confronto dos resultados do modelo com os dados empíricos. Definimos o componente cíclico como a diferença em relação a uma tendência quadrática em uma frequência anual.

No caso do Brasil, a frequência mensal é muito utilizada. Sucessivos "pacotes econômicos" e mudanças na metodologia de cálculo das séries, tornam difícil obter séries longas utilizáveis. Para não reduzir excessivamente o número de observações, séries mensais são utilizadas.

### O componente cíclico das séries

Há várias alternativas para extrair o componente cíclico das séries. Vamos seguir o exemplo principal de [Uribe & Schmitt-Grohé](https://press.princeton.edu/titles/11032.html) e utilizar uma tendência quadrática. 

#### A decomposição ciclo-tendência

Decompomos a série observada $(y_t)$ em dois componentes não-observados: a tendência ou componente secular $(y_t^S)$ e o ciclo $(y_t^C)$:

$ y_t = y_t^S + y_t^C $

O procedimento para estimar os dois componentes é bem simples. Basta estimar uma regressão tendo como variável dependente cada uma das séries de dados - portanto, quatro regressões. As variáveis independentes serão a constante, uma tendência linear $(t)$ e uma tendência quadrática $(t^2)$. O erro da regressão correponde ao componente da série $(y_t)$ não explicado pela tendência $(y_t^S)$ e, portanto, é o componente cíclico $(y_t^C)$. Ou seja:

$ y_t = \alpha_0 + \alpha_1.t + \alpha_2.t^2 + \epsilon_t $

$ y_t^S = \hat{\alpha}_0 + \hat{\alpha}_1.t + \hat{\alpha}_1.t^2 $

$ y_t^C = y_t - y_t^S = y_t - (\hat{\alpha}_0 + \hat{\alpha}_1.t + \hat{\alpha}_1.t^2) $


#### O pacote para a estimação

Para estimarmos as quatro regressões, usaremos o pacote [MultivariateStats](https://multivariatestatsjl.readthedocs.io/en/stable/index.html) com o procedimento [llsq](https://multivariatestatsjl.readthedocs.io/en/stable/lreg.html) (linear least squares).

```
Pkg.add("MultivariateStats")
import MultivariateStats: llsq
```

A sintaxe para o comando de regressão é:

```
a = llsq(X, y; bias=false)
```

a: vetor com os parâmetros estimados. X: matriz com os regressores. y: vetor com a variável dependente. O comando *bias=false* está explicado na documentação. 

#### As variáveis para a estimação

Precisamos de um vetor constante e das tendência linear e quadrática. 

(i) Medir o tamanho das séries e armazernar o valor na variável *T*. Para isso, utilizamos o comando *length*. Consulte a documentação sobre [Arrays](https://docs.julialang.org/en/v0.6.2/stdlib/arrays/). (Se as séries tivessem tamanhos diferentes seria necessário fazer isso para cada uma delas.) 

```
T=length(gbr);
```

(ii) Construir a série de anos para os eixos horizontais. O comando [LinRange](https://docs.julialang.org/en/v1/base/collections/index.html) constrói a série.

```
ano = LinRange(1960,2017,T); 
```

Terceiro, as tendências linear e quadrática para os regressores.

```
time = LinRange(1,T,T);
time2 = time.^2;
X = [ones(T) time time2];
```

Para extrair apenas uma tendência linear, basta subtituir a matriz de regressores por:

```
X = [ones(T) time];
```

Finalmente, vamos estimar as regressões e construir o componente cíclico do *PIB* brasileiro, conforme indicado acima. Os demais componentes cíclico são calculados de maneira semelhante.

```
y = cbr;
a = llsq(X, y; bias=false);
gdpargs = X * a; 
gdpargc = gdparg - gdpargs;
```

````julia
# Leitura e transformação dos dados

#Leitura dos dados
dados = readxlsheet("WDI01.xlsx","Data") ;

# Criação das variáveis 
gbr = dados[:,1];
gmx = dados[:,2];
cbr = gbr.*dados[:,3];
cmx = gmx.*dados[:,4];

# Transformação logarítmica
gbr = log.(gbr);
gmx = log.(gmx);
cbr = log.(cbr);
cmx = log.(cmx);
````

````julia
# Séries para estimação

T=length(gbr);
ano = LinRange(1960,2017,T); 
time = LinRange(1,58,58);
time2 = time.^2;
X = [ones(58) time time2];   # tendência quadrática
# X = [ones(58) time];       # tendência linear
````

````julia
# Regressões

y = gbr;
a = llsq(X, y; bias=false);
gbrs = X * a; 
gbrc = gbr - gbrs;

y = gmx;
a = llsq(X, y; bias=false);
gmxs = X * a; 
gmxc = gmx - gmxs;

y = cbr;
a = llsq(X, y; bias=false);
cbrs = X * a; 
cbrc = cbr - cbrs;

y = cmx;
a = llsq(X, y; bias=false);
cmxs = X * a; 
cmxc = cmx - cmxs;
````

````julia
# Gráficos

p1 = plot(ano, [gbr gbrs], title="BRA-PIB")

p2 = plot(ano, gbrc, title="BRA-PIB-ciclo") 

p3 = plot(ano, [cbr cbrs], title="BRA-Con")

p4 = plot(ano, cbrc, title="BRA-Con-ciclo")

p5 = plot(ano, [gmx gmxs], title="MEX-PIB")

p6 = plot(ano, gmxc,  title="MEX-PIB-ciclo") 

p7 = plot(ano, [cmx cmxs], title="MEX-Con")

p8 = plot(ano, cmxc, title="MEX-Con-ciclo")

plot(p1,p2,p3,p4,p5,p6,p7,p8, layout=(4,2), legend=false) 
#plot(p1,p2,p3,p4,p5,p6,p7,p8, layout=(4,2), legend=false, fmt = :png, size = (800, 600)) 
````

````julia
# Regressões - tendência linear

X1 = [ones(58) time];   # tendência quadrática

y = gbr;
a = llsq(X1, y; bias=false);
gbrs1 = X1 * a; 
gbrc1 = gbr - gbrs1;

y = gmx;
a = llsq(X1, y; bias=false);
gmxs1 = X1 * a; 
gmxc1 = gmx - gmxs1;

p9 = plot(ano, [gbrc gbrc1], title="BRA-PIB-ciclo") 
p10 = plot(ano, [gmxc gmxc1], title="MEX-PIB-ciclo") 
plot(p9,p10, layout=(1,2), label=["quad", "lin"], fmt = :png, size = (800, 400)) 
````

````julia
# Estatísticas

# Desvios padrão

[std(gbrc) std(cbrc)/std(gbrc);
 std(gmxc) std(cmxc)/std(gmxc)]  

````

````julia
# Correlações contemporâneas

x = cor([gbrc cbrc])     # Matriz (2x2)
y = cor([gmxc cmxc])     # Matriz (2x2)

[x[1,2] y[1,2]]          # As corrrelações em cada matriz
````

Persistência

Para calcular a autocorrelação de primeira ordem de uma série, precisamos dos seus valores contemporâneos $(y_t)$ e dos seus valores defasados $(y_{t_1})$. Por exemplo, para o *PIB* brasileiro, teremos: $y_t=gbrc[2:58]$ e $y_{t-1}=gbrc[1:57]$.

````julia
w = cor([gbrc[2:58] gbrc[1:57]])     # Matriz (2x2)
x = cor([gmxc[2:58] gmxc[1:57]])     # Matriz (2x2)
y = cor([cbrc[2:58] cbrc[1:57]])     # Matriz (2x2)
z = cor([cmxc[2:58] cmxc[1:57]])     # Matriz (2x2)

[w[1,2] x[1,2] y[1,2] z[1,2]]        # As corrrelações em cada matriz
````
