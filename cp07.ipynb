## Introdução

Nosso modelo básico, sem oferta de trabalho e sem bens de capital, supõe que a renda dos indivíduos seja exógena e que a poupança sirva apenas para transferir recursos ao longo do tempo. Neste capítulo vamos incluir investimentos produtivos no modelo. Especificamente, vamos supor que a renda em cada período dependa do estoque de capital existente naquele período. O indivíduo passa a poder utilizar seus recursos de três maneiras: consumo $(C)$, ativos financeiros $(A)$ e bens de capital $(K)$. Para simplificar, vamos supor que a oferta de trabalho seja fixa – e, portanto, não aparecerá no modelo. 

Sempre que mudamos algum aspecto da nossa economia simplificada, precisamos adaptar o modelo, refazer a otimização e recalcular o equilíbrio. Para isso, é útil ter um modelo básico robusto que possa ser facilmente adaptado para responder perguntas adicionais.

## O problema do indivíduo

Nossa economia simplificada agora inclui a possibilidade de investimentoem bens de capital. Precisamos, portanto, adaptar nosso modelo matemático e tornar a resolvê-lo. Vamos reescrever nosso modelo simplificado, começando pelo problema resolvido pelo agente individual. 

### O modelo

O problema do indivíduo com renda exógena era:

$ \left\{ \begin{array}{l l}	
\underset{(C_1,C_2,A_1)}{max} U(C_1,C_2)  = ln(C_1) + \cfrac{ln(C_2)}{1+\rho}\\
     Y_1 = C_1 + A_1 \\
     Y_2 + (1+r)A_1 = C_2  \\
    \end{array}\right. $

Com a possibilidade de investimento em bens de capital, as modificações necessárias são a inclusão: (i) dos gastos em investimento $(I)$ nas restrições orçamentárias; (ii) de uma função de produção que dependa do estoque de capital $(K)$; (iii) de uma equação de acumulação relacionando os estoques de capital em períodos sucessivos; (iv) de uma condição sobre o estoque de capital final; (v) dos investimentos e dos estoques de capital entre as variáveis de controle.

$ \left\{ \begin{array}{l l}	
\underset{X}{max} \; U(C_1,C_2)  = ln(C_1) + \cfrac{ln(C_2)}{1+\rho}\\
    Y_1 = A_1 + C_1 + I_1 \\
    Y_2 + (1+r)A_1 = C_2 + I_2  + A_2  \\
    Y_1 = Z_1K_1^\alpha  \\
    Y_2 = Z_2K_2^\alpha  \\
    K_2 = (1-\delta)K_1 + I_1 \\
    K_3 = (1-\delta)K_2 + I_2  \\
    A_2= 0 \, , \,\,\, K_3 = K^F  \\
    \\
    X = (C_1,C_2,I_1,I_2,A_1,A_2,K_2,K_3,Y_1,Y_2 )   \\
    \end{array}\right. $

$X$ representa o vetor de variáveis de controle. 

**Comentário**

(i) A função de utilidade não é afetada pelas novas possibilidades de investimento, mas as restrições o são e a lista de variáveis de controle agora inclui os estoques de capital para o segundo e o terceiro períodos. 

(ii) O estoque de capital só se torna produtivo um período após a realização do investimento. Essa hipótese captura o fato de que a maioria dos investimentos não fica pronta imediatamente. Mas, se cada período for bastante longo, alguns investimentos poderiam ficar prontos dentro do próprio período em que são realizados. 

O estoque de capital do primeiro período não aparece entre as variáveis de controle, pois ele foi escolhido “ontem”.

(iii) É comum considerar que $Z$ representa o nível da tecnologia. Na verdade, essa variável tem uma interpretação mais ampla, representando qualquer elemento que altere a produtividade total dos fatores de produção: aperfeiçoamentos de instituições como a justiça, melhorias na segurança dos contratos, na infraestrutura de transportes, greves, inundações etc...

(iv) $K^F$ é posição final (exógena) do estoque de capital que pode assumir várias formas.

Se, após serem utilizados, os bens de capital puderem ser transformados em bens de consumo: $K^F \ge 0$. Como não vale a pena experdiçar nada, $K_3^*=0$ e podemos simplificar e escrever $K^F=0$. Um argumento semelhante indica que se os bens de capital não puderem ser reconvertidos em bens de consumo: $K^F=K_2$. Utilizaremos essa última hipótese que parece mais razoável.

(v) A função de produção é um caso particular da Cobb-Douglas. Incluindo as horas trabalhadas $(H)$ na função de produção:

$ Y= ZK^\alpha H^{(1-\alpha)}  \,\,  , \,\,  0<\alpha<1 $

Fazendo $H=1$, obtemos a função de produção acima.

### As condições de ótimo

Vamos simplificar um pouco o problema para começar a resolvê-lo. Fazendo algumas substituições óbvias (<a id='voltar1'></a> [Exercício resolvido 1](#Link1)), obtemos:

$ \left\{ \begin{array}{l l}	
\underset{X}{max} \; U(C_1,C_2)  = ln(C_1) + \cfrac{ln(C_2)}{1+\rho}\\
    Z_1K_1^α  + \cfrac{Z_2K_2^α}{1+r} = C_1 + K_2 - (1-\delta)K_1 +
        \cfrac{C_2 + K^F - (1-\delta)K_2}{1+r}\\
    X = (C_1,C_2,K_2)   \\
    \end{array}\right. $
    
O indivíduo tem três variáveis de escolha e uma restrição e, portanto, apenas duas escolhas independentes. Em relação ao problema original, a cada simplificação, eliminamos uma restrição e uma variável de escolha e, portanto, os graus de liberdade do indivíduo não mudaram. Eles correspondem a duas escolhas economicamente relevantes: (i) quanto vai consumir e poupar no primeiro período e (ii) como vai alocar sua poupança entre ativos financeiros e bens de capital.

O Lagrangeano do problema é:

$ \left\{ \begin{array}{l l}	
\underset{X}{max} \; U(C_1,C_2)  = ln(C_1) + \cfrac{ln(C_2)}{1+\rho} + 
    \lambda \left[ Z_1K_1^α  + \cfrac{Z_2K_2^α}{1+r} - C_1 - K_2 + (1-\delta)K_1 -
        \cfrac{C_2 + K^F - (1-\delta)K_2}{1+r} \right]\\
    X = (C_1,C_2,K_2,\lambda)   \\
    \end{array}\right. $
    
E as condições de primeira ordem são:

$ \left\{ \begin{array}{l l}
(i)\,\, \cfrac{\partial L}{\partial C_1} = \cfrac{1}{C_1} - \lambda = 0  \\
(ii)\,\, \cfrac{\partial L}{\partial C_2} = \cfrac{1}{1+\rho} \cfrac{1}{C_2} - \cfrac{\lambda}{1+r} = 0 \\
(iii) \,\, \cfrac{\partial L}{\partial K_2} = \alpha Z_2K_2^{\alpha-1} - (r+\delta) = 0      \\
(iv)\,\, \cfrac{\partial L}{\partial \lambda} = Z_1K_1^α  + \cfrac{Z_2K_2^α}{1+r} - C_1 - K_2 + (1-\delta)K_1 -
        \cfrac{C_2 + K^F - (1-\delta)K_2}{1+r} = 0 \\
\end{array}\right.$

**Comentário**

(i) A primeira condição indica que a escolha ótima do consumo presente depende apenas da utilidade marginal da riqueza. (Lembre que o multiplicador $\lambda$ representa a utilidade marginal da riqueza.) Quando o indivíduo fica mais rico (pobre), a utlidade marginal da riqueza diminui (aumenta) e ele aumenta (reduz) o consumo corrente, reduzindo a utilidade marginal do mesmo.

A segunda condição tem uma interpretação equivalente para o consumo do segundo período, desde que as utilidades marginais do consumo e da riqueza sejam trazidas a valor presente - pela taxa de impaciência e de juros, respectivamente.

Existe uma aparente assimetria entre as escolhas de consumo ótimo presente e futuro, já que a segunda condição envolve as taxas de desconto intertemporal  e a primeira não. Mas essa assimetria é apenas aparente. Se o problema tivesse sido escrito em termos de unidades do segundo período, as taxas intertemporais apareceriam na primeira condição.

(ii) A terceira condição não envolve o multiplicador, indicando que a decisão de investimento não depende da riqueza do indivíduo. Um investimento é rentável ou não independente do investidor ser rico ou pobre. (Voltaremos a discutir esse ponto.)

♠



O <a id='voltar2'></a> [Exercício resolvido 2](#Link2) mostra que as condições de primeira ordem implicam três expressões de fácil interpretação econômica:

$ \left\{ \begin{array}{l l}
    r=\alpha Z_2K_2^{\alpha-1} - \delta  \\
    Z_1K_1^α  + \cfrac{Z_2K_2^α}{1+r} = C_1 + \overbrace{K_2 - (1-\delta)K_1}^{I_1} +
        \cfrac{C_2 + \overbrace{K^F - (1-\delta)K_2}^{I_2}}{1+r}\\
    \cfrac{C_2}{C_1}=\cfrac{1+r}{1+\rho}   \\
\end{array}\right. $

**Comentário**

(i) A primeira condição mostra que a escolha ótima do estoque de capital do segundo período é tal que sua produtividade marginal deve ser igual à taxa de juros; ou seja, o retorno de um projeto deve ser igual ao seu custo. Essa equação não envolve nenhuma variável relacionada às preferências do indivíduo, apenas elementos financeiros e tecnológicos. Esse ponto é importante e será discutido a seguir. 

A segunda condição é a restrição orçamentária intertemporal. O valor presente dos recursos (lado esquerdo) deve ser igual ao valor presente dos gastos (lado direito).

A última equação é a equação de Euler, que determina os consumos ótimos. Vemos que, dada a taxa de juros, o perfil de consumo desejado não é alterado porque o indivíduo tem novas possiblidades de investimento.

(ii) Note que as condições de ótimo podem ser resolvidas de forma recursiva. A primeira equação envolve uma única variável endógena, $K_2$, e, portanto, determina valor ótimo do do estoque de capital do segundo período ($ K_2^*$).  Substituindo na  restrição orçamentária intertemporal:

$   Z_1K_1^α  + \cfrac{Z_2K_2^{*α}}{1+r} - K_2^* + (1-\delta)K_1 -
        \cfrac{K^F - (1-\delta)K_2^*}{1+r} = W^* = C_1  + \cfrac{1}{1+r}C_2 $
        
A última igualdade mostra que, uma vez calculado o capital ótimo, estamos de volta no problema de escolha do perfil ótimo de consumo com recursos exógenos. Para ser claro: a riqueza do indivíduo é endógena nesse problema. Mas, nesse caso específico, a solução do problema pode ser dividida em duas partes. Na primeira parte, o indivíduo escolhe o estoque ótimo de capital para o segundo período. Uma vez que o capital tenha sido escolhido, a riqueza fica determinada. Na segunda parte da solução, dado o valor da riqueza, o agente escolhe como alocá-la entre consumo corrente e futuro, combinado a equação de Euler como a restrição orçamentária..

(iii) Reorganizando a restrição orçamentária intertemporal, vemos os recursos de que o indivíduo dispõe para consumir no segundo período:

$ C_2=(1+r)[\overbrace{(\underbrace{Z_1K_1^α}_{Y_1} - C_1 - (\underbrace{K_2-(1-\delta)K_1}_{I_1})}^{A_1}] 
+ \underbrace{Z_2K_1^α}_{Y_2} + \underbrace{(1-\delta)K_2-K^F}_{-I_2}  $

As fontes de recursos para o indivíduo consumir no segundo período são: (i) o rendimento da poupança financeira do primeiro período, $(A_1)$; (ii) a produção do segundo período, $(Y_2)$; (iii) o estoque de capital que é desinvestido, $((1-\delta)K_2-K^F))$.

(iv) Vimos que o estoque ótimo de capital não depende de nenhuma variável relacionada às preferências do indivíduo, apenas elementos financeiros e tecnológicos. Esse é um ponto importante. 

Considere dois indivíduos idênticos, exceto por um ser mais impaciente que o outro: $ρ^A>ρ^B$. Qual dos dois investirá mais? O modelo indica que ambos investirão igualmente, pois a impaciência não entra na primeira equação que, sozinha, determina a escolha ótima do estoque de capital: $I_1^A=I_1^B,K_2^A=K_2^B$.

Considere, agora, dois indivíduos idênticos, exceto por um ter um estoque maior de capital inicial $K_1^A=K_1^B$? Qual dos dois investirá mais? Qual dos dois investirá terá mais capital no segundo período? O modelo indica que ambos terão o mesmo estoque de capital no segundo período $(K_2^A=K_2^B )$ e, portanto, o que tem o maior capital inicial investirá menos $ (I_1^A=I_1^B ) $. 

O que está acontecendo? O ponto fundamental é o acesso irrestrito ao mercado de capitais. 

A regra de investimento diz que a pessoa deve investir até que a produtividade marginal do capital (retorno do projeto) seja igual à taxa de juros (custo do projeto). Abrir mão de qualquer oportunidade de investimento cujo retorno seja superior ao custo é perder dinheiro. Se você não tem os recursos necessários para investir, peça emprestado; se a produtividade marginal é maior que a taxa de juros, você ficará mais rico. Se você é muito impaciente, invista em todos os projetos em que o retorno seja superior ao custo e peça emprestado para financiar seu consumo corrente. Desde que você tenha acesso irrestrito ao mercado de capitais, essa é a estratégia ideal. 

E se, como acontece frequentemente na prática, você não conseguir todo o financiamento que deseja? Veremos isso na próxima seção. Mas, o que aprendemos sobre essa situação pouco realista não é inútil. Primeiro, as pessoas tentarão chegar o mais próximo possível da situação ideal descrita acima. Segundo, instituições que facilitem o acesso ao crédito, tendem a aumentar o bem-estar das pessoas. Dificultar a cobrança de dívidas pode ser bom para quem já obteve crédito, mas é ruim para aqueles que ainda precisam de crédito.

## Conclusão

Este capítulo discutiu o problema de alocação intertemporal de recursos de um consumidor que tem acesso a investimentos financeiros e em capital físico. Isso significa que a renda (do segundo período) passa a ser endógena. A renda também seria endógena caso o indivíduo pudesse escolher o número de horas a trabalhar, como vermos em um próximo capítulo.

Vimos que a condição para a escolha do perfil ótimo de consumo ao longo do tempo não muda e a equação de Euler continua valendo. Esse é um resultado bastante robusto que aparecerá nos diversos modelos que veremos.

Em segundo lugar, vimos que, na presença de um mercado de capitais ideal, o consumidor pode separar sua decisão de investimento da sua decisão de consumo. O indivíduo investirá até que a produtividade marginal do capital se iguale ao custo do capital – a taxa de juros –, financiando suas despesas no mercado de capitais. Esse resultado não vale se a hipótese de um mercado de capitais ideal não for uma boa aproximação para o problema que estivermos estudando. Em termos agregados, o resultado não vale em uma autarquia, como veremos no próximo capítulo.

```julia
# Importação dos pacotes

Pkg.add("ExcelReaders")
import ExcelReaders: readxlsheet

Pkg.add("MultivariateStats") 
import MultivariateStats: llsq

Pkg.add("Plots")
import Plots: plot

Pkg.add("StatsBase")
import StatsBase: std, cor
```

```julia
# Leitura e transformação dos dados

dados = readxlsheet("WDI02.xlsx","Data") ;

gbr=dados[:,1];
gmx=dados[:,2];
ibr=gbr.*dados[:,3];
imx=gmx.*dados[:,4];
gbr=log.(gbr);
gmx=log.(gmx);
ibr=log.(ibr);
imx=log.(imx);
```

```julia
# Séries para estimação

T=length(gbr);
ano = LinRange(1960,2017,T); 
time = LinRange(1,58,58);
time2 = time.^2;
X = [ones(58) time time2];   # tendência quadrática
# X = [ones(58) time];       # tendência linear
```

```julia
# Regressões

y = gbr;
a = llsq(X, y; bias=false);
gbrs = X * a; 
gbrc = gbr - gbrs;

y = gmx;
a = llsq(X, y; bias=false);
gmxs = X * a; 
gmxc = gmx - gmxs;

y = ibr;
a = llsq(X, y; bias=false);
ibrs = X * a; 
ibrc = ibr - ibrs;

y = imx;
a = llsq(X, y; bias=false);
imxs = X * a; 
imxc = imx - imxs;
```

```julia
# Gráficos

p1 = plot(ano, [gbr gbrs], title="BRA-PIB")

p2 = plot(ano, gbrc, title="BRA-PIB-ciclo") 

p3 = plot(ano, [ibr ibrs], title="BRA-Inv")

p4 = plot(ano, ibrc, title="BRA-Inv-ciclo")

p5 = plot(ano, [gmx gmxs], title="MEX-PIB")

p6 = plot(ano, gmxc,  title="MEX-PIB-ciclo") 

p7 = plot(ano, [imx imxs], title="MEX-Inv")

p8 = plot(ano, imxc, title="MEX-Inv-ciclo")

#plot(p1,p2,p3,p4,p5,p6,p7,p8, layout=(4,2), legend=false) 
plot(p1,p2,p3,p4,p5,p6,p7,p8, layout=(4,2), legend=false, fmt = :png, size = (800, 600)) 
```

```julia
# Estatísticas

# Desvios padrão

[std(gbrc) std(ibrc)/std(gbrc);
 std(gmxc) std(imxc)/std(gmxc)]   
```

```julia
# Correlações contemporâneas
x = cor([gbrc ibrc])     # Matriz (2x2)
y = cor([gmxc imxc])     # Matriz (2x2)
[x[1,2] y[1,2]]          # As corrrelações em cada matriz
```
